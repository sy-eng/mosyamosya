```python
import pandas as pd
import numpy as np
import sys

class Item:
    def __init__(self):
        # Evaluate the weight
        self.weight = 0
        self.path = []

    def __init__(self, weight, path):
        # Evaluate the weight
        self.weight = weight
        self.path = path

def printDpMatrix(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.weight, end = " ")
        print(" ")
        
    print(" ")
    
def printDpMatrixPath(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.path, end = " ")
        print(" ")
        
    print(" ")

# STEP 0 - Initialization
dataFilePath = "./knapsack2.txt"

# The format is different from 4.4
# The row 1 has only 1 element. Because of this, the format for the "price"  must be "float64"  
data = pd.read_csv(dataFilePath,encoding="sjis", comment="#", sep = ",", names = ["weight", "price"], dtype = {'weight':'float64', 'price':'float64'})

items = np.array(data.loc[1:].values)
weightCap = int(data.loc[0].values[0])
itemNum = len(items)

priceList = np.array(items).T[1]
print(weightCap)

#Calculate the sum of price
priceSum = int(sum(priceList))
print(priceSum)
print(data.loc[1:])
#print(items)

# In this program, this matrix is n(Psum+1)
# But, this can be 2(Psum+1). 
dpMatrix = []

# STEP 1
tmpMatrix = [Item(sys.maxsize, []) for i in range(priceSum + 1)]
tmpMatrix[0].weight = 0
tmpMatrix[int(items[0][1])].weight = items[0][0]
tmpMatrix[int(items[0][1])].path = [1]

dpMatrix.append(tmpMatrix)

# STEP 2
# This is the same with 4.4
while itemNum > len(dpMatrix):
    index = len(dpMatrix)
    
    tmpMatrix = []
    
    #STEP 3
    for i in range(priceSum + 1):
        if i < int(items[index][1]):
            tmpMatrix.append(Item(dpMatrix[index - 1][i].weight, dpMatrix[index - 1][i].path))
        else:
            prevWeight = dpMatrix[index - 1][i].weight
            addedWeight = dpMatrix[index - 1][i - int(items[index][1])].weight + items[index][0]
            
            if  prevWeight > addedWeight:
                tmpMatrix.append(Item(addedWeight, dpMatrix[index - 1][i - int(items[index][1])].path + [index + 1]))
            else:
                tmpMatrix.append(Item(prevWeight, dpMatrix[index - 1][i].path))
    
    dpMatrix.append(tmpMatrix)
    
    #printDpMatrix(dpMatrix)

# Show the result.
printDpMatrix(dpMatrix)
printDpMatrixPath(dpMatrix)

for i in range(priceSum + 1):
    if dpMatrix[-1][priceSum - i].weight <= weightCap:
        print("Max value : ", priceSum - i)
        print("Items : ", dpMatrix[-1][priceSum - i].path)
        break
```
