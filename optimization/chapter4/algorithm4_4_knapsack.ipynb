```python
import pandas as pd
import numpy as np

class Item:
    def __init__(self):
        self.price = 0
        self.path = []

    def __init__(self, price, path):
        self.price = price
        self.path = path

def printDpMatrix(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.price, end = " ")
        print(" ")
        
    print(" ")
    
def printDpMatrixPath(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.path, end = " ")
        print(" ")
        
    print(" ")

    # STEP 0 - Initialization
dataFilePath = "./knapsack.txt"

data = pd.read_csv(dataFilePath,encoding="sjis", comment="#", sep = ",", names = ["weight", "price"], dtype = {'weight':'uint64', 'price':'float64'})

items = np.array(data.loc[1:].values)
capacity = int(data.loc[0].values[0])
itemNum = len(items)

print(capacity)
print(data.loc[1:])
#print(items)

# In this program, this matrix is n(C+1)
# But, this can be 2(C+1). 
dpMatrix = []

#STEP 1
tmpMatrix = []
for i in range(capacity + 1):
    if i < int(items[0][0]):
        tmpMatrix.append(Item(0, []))
    else:
        tmpMatrix.append(Item(items[0][1], [1]))
        
dpMatrix.append(tmpMatrix)

#printDpMatrix(dpMatrix)

#STEP 2
while itemNum > len(dpMatrix):
    index = len(dpMatrix)
    
    tmpMatrix = []
    
    #STEP 3
    for i in range(capacity + 1):
        if i < int(items[index][0]):
            tmpMatrix.append(Item(dpMatrix[index - 1][i].price, dpMatrix[index - 1][i].path))
        else:
            prevPrice = dpMatrix[index - 1][i].price
            addedPrice = dpMatrix[index - 1][i - int(items[index][0])].price + items[index][1]
            
            if  prevPrice < addedPrice:
                tmpMatrix.append(Item(addedPrice, dpMatrix[index - 1][i - int(items[index][0])].path + [index + 1]))
            else:
                tmpMatrix.append(Item(prevPrice, dpMatrix[index - 1][i].path))
    
    dpMatrix.append(tmpMatrix)
    
    #printDpMatrix(dpMatrix)

# Show the result.
printDpMatrix(dpMatrix)
printDpMatrixPath(dpMatrix)

print("Max value : ", dpMatrix[-1][-1].price)
print("Items in the knapsack : ", dpMatrix[-1][-1].path)
```
