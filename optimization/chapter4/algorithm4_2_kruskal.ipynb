```python
import pandas as pd
import numpy as np
from optimizationLib import Node

class QuickSort:
    # The data is not checked.
    # The pivot is the first element.
    # Here is O(N log N) in the best case or on average
    # Here is O(N log N) in the worst case
    def execute(self, nodes):
        
        if len(nodes) == 0:
            return []
        
        if len(nodes) == 1:
            return nodes
        
        smaller = []
        larger = []
        
        pivot = [nodes[0]]
        
        # Here can be O(N)
        for n in nodes[1:]:
            if(n.value <= nodes[0].value):
                smaller.append(n)
            else:
                larger.append(n)

        # This recursive process is O(log N) in the best case or on average
        # This recursive process is O(N^2) in the worst case
        return self.execute(smaller) + pivot + self.execute(larger)
    
# STEP 0 - Initialization
dataFilePath = "./kruskal.txt"

data = pd.read_csv(dataFilePath,encoding="sjis", comment="#", sep = ",", names = ["index1", "index2", "data"])

ps = np.array(data.loc[:].values.tolist())

maxId = 0
nodes = []
fs = []
fsElem = []
ts = []

for p in ps:
    nodes.append(Node(p[2], [p[0], p[1]]))
    
    if maxId < p[0]:
        maxId = p[0]
    
    if maxId < p[1]:
        maxId = p[1]

for i in range(maxId):
    fs.append(i + 1)
    fsElem.append([i + 1])

#print(fs)
#print(fsElem)

#STEP 1 - Sort the edges
qs = QuickSort()
nodes = qs.execute(nodes)

for n in nodes:
    print(n.index, " : ", n.value)

# Here is O(|E|) = O(|V|^2)
for n in nodes:
    #STEP 2 
    if len(ts) == (maxId - 1):
        break
        
    #STEP 3
    #These process are done |V| times.
    if fs[n.index[0] - 1] != fs[n.index[1] - 1]:
        ts.append(n)
        
        if len(fsElem[fs[n.index[0] - 1] - 1]) < len(fsElem[fs[n.index[1] - 1] - 1]):
            attachedIndex = 1
            removedIndex = 0
        else:
            attachedIndex = 0
            removedIndex = 1
        
        fsRemoveIndex = fs[n.index[removedIndex] - 1] - 1
        
        # Here is O(|V|)
        for e in fsElem[fs[n.index[removedIndex] - 1] - 1]:
            fs[e - 1] = fs[fs[n.index[attachedIndex] - 1] - 1]
            fsElem[fs[n.index[attachedIndex] - 1] - 1].append(e)
        
        fsElem[fsRemoveIndex] = []
        
        print(fs)
        #print(fsElem)

print("Selected edges")
for t in ts:
    print(t.index)
```
