```python
from optimizationLib import Node

B = 6

class Item:
    def __init__(self, value, path):
        self.value = value
        self.path = path

def f1(x1):
    return 10 * abs(x1 - 1)

def f2(x2):
    return x2**2

def f3(x3):
    return 2 * (x3 - 1)**2

def f4(x4):
    return 60 / abs(2 * x4 - 9)

def printDpMatrix(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.value, end = " ")
        print(" ")
        
    print(" ")
    
def printDpMatrixPath(dpMatrix):
    for mRow in dpMatrix:
        for m in mRow:
            print(m.path, end = " ")
        print(" ")
        
    print(" ")

funcs = [f1, f2, f3, f4]

dpMatrix = []

n = len(funcs)

# STEP 1
tmpMatrix = []

for i in range(B + 1):
    tmpMatrix.append(Item(funcs[0](i), [i]))

dpMatrix.append(tmpMatrix)

k = 1
#k = len(dpMatrix)

# STEP 2
while k < n:
    
    # STEP 3
    tmpMatrix = []
    
    # Here is O(B^2)
    for i in range(B + 1):
        # Here is O(B)
        maxOutput = dpMatrix[k - 1][i].value + funcs[k](0)
        maxIndex = 0
        
        for j in range(1, i + 1):
            tmpValue = dpMatrix[k - 1][i - j].value + funcs[k](j)
            
            if tmpValue > maxOutput:
                maxOutput = tmpValue
                maxIndex = j
                
        tmpMatrix.append(Item(maxOutput, dpMatrix[k-1][i - maxIndex].path + [maxIndex]))

    dpMatrix.append(tmpMatrix)

    # Here is STEP 2
    k = k + 1

# Show the result.
printDpMatrix(dpMatrix)
printDpMatrixPath(dpMatrix)

maxResult = dpMatrix[-1][0].value
maxIndex = 0

for i in range(1, len(dpMatrix[-1])):
    if maxResult < dpMatrix[-1][i].value:
        maxResult = dpMatrix[-1][i].value
        maxIndex = i
        
print(maxResult, " : ", dpMatrix[-1][maxIndex].path)
```
