完全ユニモジュラ行列の十分条件について証明をまとめておく。

[このページ](http://dopal.cs.uec.ac.jp/okamotoy/lect/2014/fdopt/handout07.pdf)のスライド38-43の内容を拡張することにより、2部グラフだけでなく、本にある、十分条件を示すことができる。

十分条件は本にあるように整数行列${\bf A} = (a_{ij}) \in \mathbb{Z}^{m \times n}$が以下の条件を満たすことである。

1. 任意の要素$a_{ij}$は0,1,-1のいずれかの値を取る。
1. 任意の列jは高々2つの非零の要素を持つ。($\sum_{i=1}^m |a_{ij}| \leq 2$)
1. 任意の列jに対して、$\sum_{i \in M_1} a_{ij} - \sum_{i \in M_2} a_{ij} \in \{-1, 0, 1\}$となる、$M=\{1, \cdots, m\}$の分割$(M_1, M_2)$が存在する。

この3つの条件のうち、1については、任意の正方部分行列で成立するため、$a_{ij}$のみを要素とする、$1 \times 1$正方部分行列でも成立しなければならない。よって、1は十分条件の1つになる。

2, 3について、考える。

1の条件より、任意の$1 \times 1$正方部分行列で行列式が、-1, 0, 1になっていることが言えている。
更に、$1 \times 1$正方部分行列では2, 3の条件も満たしている。

帰納法を利用し、$k = l - 1$までで、 十分条件を満たしていれば、${\bf A}の$任意の$k \times k$正方部分行列で行列式が、-1, 0, 1になっていることが言えているときに、
$k = l$のときも任意の$k \times k$正方部分行列で行列式が、-1, 0, 1になっていることを言う。

$k = l$のときに以下のように場合分けをする。

1. 0のみの要素しか持たない列がある。
1. 上記でないとき、1個だけ、非零の要素を持つ列がある。
1. すべての列が2個の非零の要素を持つ列である。

このように場合分けできるのは2の条件による。

1の0のみの要素しか持たない列があるときは、明らかに、行列式が0になる。

2の0のみの要素を持つ列がないとき、1個だけ、非零の要素を持つ列があるときは、それが1だとすると、適当な置換$\sigma$を行うことで部分行列${\bf B}$の行列式は以下のようになる。

$$
det {\bf B} = sgn(\sigma)
\begin{vmatrix}
1 & {\bf b}\\
{\bf 0} & {\bf B}'\\
\end{vmatrix}
= sgn(\sigma) det {\bf B}' \in \{-1, 0, 1\}
$$
ここで、${\bf B},{\bf B}'$ともに、${\bf A}$の部分行列である。${\bf B}'$は$(l-1) \times (l - 1)$の正方部分行列で、帰納法の仮定から、
行列式は{-1, 0, 1}のいずれかだとわかっている。また、$sgn(\sigma)$は-1か1であり、非零の要素が-1のときは-1をかければよく、部分行列の
行列式が{-1, 0, 1}になるという結論に変わりはない。

3のすべての列が2個の非零の要素を持つ列であるときは、行列式が0になることを言う。

線形代数の結果から、${\bf B}の$行列式が0でなければ、任意のベクトル${\bf x} \neq {\bf 0}$に対して、${\bf x^T B} \neq {\bf 0}^T$だと言えている。

(${\bf B}の$行列式が0でなければ、${\bf B}$は正則で、逆行列を持つので、${\bf x^t B} = {\bf 0}^T$であれば${\bf x}^T = {\bf x^T B^{-1}} = {\bf 0}^T$)

よって、その対偶から、ある${\bf x} \neq {\bf 0}$に対して、${\bf x^T B} = {\bf 0}^T$ならば、${\bf B}$の行列式が0になる。

さて、十分条件の3より、すべての列に対して、共通の$M_1,M_2$への行の分割が存在している。その時、yの要素を$M_1$に対応する行を1,$M_2$に対応する行を-1とすると、${\bf y^T B} = (\sum_{i \in M_1} a_{ij} - \sum_{i \in M_2} a_{ij})^T$となる。

${\bf y}^T \neq {\bf 0}$であり、かつ、非零の要素は2個あるので、$\sum_{i \in M_1} a_{ij} - \sum_{i \in M_2} a_{ij}$は偶数になり、-1, 0, 1のどれかなので、0になる。

よって、${\bf y^T B} = {\bf 0}^T$となり、${\bf B}$の行列式は0となる。

上記の1,2,3の場合分け、全てで、$k = l$のときも任意の$k \times k$正方部分行列で行列式が、-1, 0, 1になっているため、1,2,3の条件を
満たしていれば、完全ユニモジュラ行列になっていると言える。
