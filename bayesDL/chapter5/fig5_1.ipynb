```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

H1 = [2, 2, 8, 8]
sigmaW2 = [1.0, 100.0, 1.0, 100.0]
inputNum = 2
xs = np.linspace(-5, 5, 1000)

fig = plt.figure()
num = 1

lineNum = 5

for h, s in zip(H1, sigmaW2):
    
    ax = fig.add_subplot(2, 2, num)
    ax.set_title('H1 : {}, Sigma : {}'.format(h, s))
    
    for line in range(lineNum):
        wh1 = []
        outputData = []

        for i in range(h):
            wh1.append(np.random.normal(loc = 0, scale = np.sqrt(s), size = inputNum))
    #    print(wh1)
        wh1 = np.array(wh1)
    
        #2層目の定数項も考えないと、関数の左右に段差がつかなさそう。
        wh2 = np.random.normal(loc = 0, scale = np.sqrt(s), size = h + 1)
    
    #    print(wh1)
    
        for x in xs:
            inputValue = np.array([x, 1])
            firstLayer = inputValue.dot(wh1.T)

    #        print(inputValue)
    #        print(firstLayer)
    
            hiddenTmp = tanh(np.array(firstLayer))
            hiddenLayer = np.append(hiddenTmp, 1.0)
            #hiddenLayer = np.append(hiddenTmp, 0.0)
        
    #        print(hiddenLayer)
        
            outputData.append(hiddenLayer.dot(wh2))
    
    #    print(outputData)
        ax.plot(xs, outputData)
    
    num = num + 1

plt.subplots_adjust(wspace=0.2, hspace=0.4)
plt.show()
```
