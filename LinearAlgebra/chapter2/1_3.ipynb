まず、$P_2$を考える。
そもそも、
\begin{eqnarray}
{P_2}^2 =
\begin{vmatrix}
0       & x_{12}\\
-x_{12} & x_0   \\
\end{vmatrix}
= {x_{12}}^2
\end{eqnarray}
になる。最初の等号はP85にある、$P_2$(もしくは$P_n$)の定義になる。よって$P_2 = x_{12}$。(符号は適当で良い)  
$P_4$を考えてみる。  
P86の6行目の式(この式はP86の2行目までで証明されている。)を考えると、$P_4$で$p = 2$, $P_2 = P_2(x_{12}) = x_{12}$に注意して、
$$
P_4 = P_4(x_{12}, x_{13}, x_{14}, x_{23}, x_{24}, x_{34}) = {x_{12}}^{(2 - 2)} P_2(x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23})
= x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23}
$$
この式は、
$$
P_4(x_{12}, x_{13}, x_{14}, x_{23}, x_{24}, x_{34}) = x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23} 
= \sum_{i_1<i_2, i_3<i_4, i_1<i_3}\epsilon 
\begin{pmatrix}
1   & 2   & 3   & 4\\
i_1 & i_2 & i_3 & i_4
\end{pmatrix}
x_{i_1i_2}x_{i_3i_4}
$$
更に、$x_{ij} = -x_{ji}$なので、  
$$
P_4(x_{12}, x_{13}, x_{14}, x_{23}, x_{24}, x_{34}) = \frac{1}{2^{2}2!}
((x_{12}x_{34} - x_{12}x_{43} + x_{34}x_{12} - x_{34}x_{21} - x_{21}x_{34} + x_{21}x_{43} - x_{43}x_{12} + x_{43}x_{21})
-(x_{13}x_{24} - x_{13}x_{42} + x_{24}x_{13} - x_{24}x_{31} - x_{31}x_{24} + x_{31}x_{42} - x_{42}x_{13} + x_{42}x_{31})
+(x_{14}x_{23} - x_{14}x_{32} + x_{23}x_{14} - x_{23}x_{41} - x_{41}x_{23} + x_{41}x_{32} - x_{32}x_{14} + x_{32}x_{41}))
= \frac{1}{2^{2}2!}\sum_{(i_1, i_2, i_3, i_4)}\epsilon 
\begin{pmatrix}
1   & 2   & 3   & 4\\
i_1 & i_2 & i_3 & i_4
\end{pmatrix}
x_{i_1i_2}x_{i_3i_4}
$$
よって、$P_4$に対して、P86の7,8行目の式が成り立っていることがわかる。  
$P_6$を考えてみる。P86の5行目から、
$$
P_6(x_{12}, x_{13}, x_{14}, x_{15}, x_{16}, x_{23}, x_{24}, x_{25}, x_{26}, x_{34}, x_{35}, x_{36}, x_{45}, x_{46}, x_{56}) = {x_{12}}^{2-3} P_4(x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23}, x_{12}x_{35} - x_{13}x_{25} + x_{15}x_{23}, x_{12}x_{36} - x_{13}x_{26} + x_{16}x_{23}, x_{12}x_{45} - x_{14}x_{25} + x_{15}x_{24}, x_{12}x_{46} - x_{14}x_{26} + x_{16}x_{24}, x_{12}x_{56} - x_{15}x_{26} + x_{16}x_{25})
$$
これを展開するとえらいことになるので、P86の9-10行目について、検討する。$P_4$の$x_{12}x_{34}$と$x_{12}x_{43}$を対応させて、
$$
(P_4のx_{12}x_{34}相当) = (x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23})(x_{12}x_{56} - x_{15}x_{26} + x_{16}x_{25})
= x_{12}x_{34}x_{12}x_{56} - (x_{12}x_{34}x_{15}x_{26} + x_{13}x_{24}x_{12}x_{56}) + (x_{12}x_{34}x_{16}x_{25} + x_{14}x_{23}x_{12}x_{56}) + (x_{13}x_{24} - x_{14}x_{23})(x_{15}x_{26} - x_{16}x_{25})
$$
$$
(P_4のx_{12}x_{43}相当) = (x_{12}x_{34} - x_{13}x_{24} + x_{14}x_{23})(x_{12}x_{65} - x_{16}x_{25} + x_{15}x_{26})
= - x_{12}x_{34}x_{12}x_{56} + (x_{12}x_{34}x_{15}x_{26} + x_{13}x_{24}x_{12}x_{56}) - (x_{12}x_{34}x_{16}x_{25} + x_{14}x_{23}x_{12}x_{56}) + (x_{13}x_{24} - x_{14}x_{23})(x_{15}x_{26} - x_{16}x_{25})
$$
$P_4$では$x_{12}x_{34}$から$x_{12}x_{43}$を引いているので、
$$
(P_4のx_{12}x_{34}相当) - (P_4のx_{12}x_{43}相当) = 2(x_{12}x_{34}x_{12}x_{56} - (x_{12}x_{34}x_{15}x_{26} + x_{13}x_{24}x_{12}x_{56}) + (x_{12}x_{34}x_{16}x_{25} + x_{14}x_{23}x_{12}x_{56}))=2x_{12}(x_{12}x_{34}x_{56} - (x_{34}x_{15}x_{26} + x_{13}x_{24}x_{56}) + (x_{34}x_{16}x_{25} + x_{14}x_{23}x_{56}))
$$
よりコンパクトにはなるが、ひとまず、カッコ内の後ろ2項を掛ける部分が2個(以上(今回は2個しかない))あれば、その項は他の項と打ち消す。  
つまり、1項目をすべてかけたものとどれかの2項目、3項目と1項目をかけた項のみが残る。
また、2項目、3項目相当の項は括弧の中に、それぞれ2項あり、これはp-1に相当する。(p-1箇所から2項目、3項目に相当する場所を1箇所選ぶ。)
置換全体に対して足し合わせるので、$P_6$の式は$P_4$の式に1項目と、2,3項目の2(=p-1)倍を入れたものになり、
$$
P_6(x_{12}, x_{13}, x_{14}, x_{15}, x_{16}, x_{23}, x_{24}, x_{25}, x_{26}, x_{34}, x_{35}, x_{36}, x_{45}, x_{46}, x_{56}) = {x_{12}}^{2-3} \frac{1}{2^{2}2!}\sum_{(i_3, i_4, i_5, i_6)}\epsilon 
\begin{pmatrix}
3   & 4   & 5   & 6\\
i_3 & i_4 & i_5 & i_6
\end{pmatrix}
((x_{12}x_{i_3i_4}x_{12}x_{i_5i_6}) - 2x_{1i_3}x_{2i_4}x_{12}x_{i_5i_6} + 2x_{1i_4}x_{2i_3}x_{12}x_{i_5i_6})
= \frac{1}{2^{2}2!}\sum_{(i_3, i_4, i_5, i_6)}\epsilon 
\begin{pmatrix}
3   & 4   & 5   & 6\\
i_3 & i_4 & i_5 & i_6
\end{pmatrix}
((x_{12}x_{i_3i_4}x_{i_5i_6}) - 2x_{1i_3}x_{2i_4}x_{i_5i_6} + 2x_{1i_4}x_{2i_3}x_{i_5i_6})
$$
和の中の2項目と、3項目は$i_3, i_4$を置換するとおなじになる。置換すると$\epsilon$の影響で-1を掛けることになるので、
$$
= \frac{1}{2^{2}2!}\sum_{(i_3, i_4, i_5, i_6)}\epsilon 
\begin{pmatrix}
3   & 4   & 5   & 6\\
i_3 & i_4 & i_5 & i_6
\end{pmatrix}
((x_{12}x_{i_3i_4}x_{i_5i_6}) - 2(p - 1)x_{1i_3}x_{2i_4}x_{i_5i_6})
= \frac{1}{2^{2}2!}\sum_{(i_3, i_4, i_5, i_6)}\epsilon 
\begin{pmatrix}
3   & 4   & 5   & 6\\
i_3 & i_4 & i_5 & i_6
\end{pmatrix}
x_{12}x_{i_3i_4}x_{i_5i_6} - \frac{1}{2^{1}1!}\sum_{(i_3, i_4, i_5, i_6)}\epsilon 
\begin{pmatrix}
3   & 4   & 5   & 6\\
i_3 & i_4 & i_5 & i_6
\end{pmatrix}
x_{1i_3}x_{2i_4}x_{i_5i_6}
$$
以上から、P_6において、
$$
P_n = \frac{1}{2^{p-1}(p-1)!}\sum_{(i_3, i_4, ..., i_n)}\epsilon 
\begin{pmatrix}
3   & 4   & \cdots & n\\
i_3 & i_4 & \cdots & i_n
\end{pmatrix}
x_{12}x_{i_3i_4}...x_{i_{n-1}i_n} - \frac{1}{2^{p-2}(p-2)!}\sum_{(i_3, i_4, ..., i_n)}\epsilon 
\begin{pmatrix}
3   & 4   & \cdots & n\\
i_3 & i_4 & \cdots & i_n
\end{pmatrix}
x_{1i_3}x_{2i_4}...x_{i_{n-1}i_n}
$$
が成立している。  
$P_n$は必ず、$x_{12}$もしくは$x_{1i_3}x_{2_i4}$を含むわけだから、上記の式と以下の式の係数を比較する。
$$
P_n = \frac{1}{2^{p}p!}\sum_{(i_1, i_2, ..., i_n)}\epsilon 
\begin{pmatrix}
1   & 1   & \cdots & n\\
i_1 & i_2 & \cdots & i_n
\end{pmatrix}
x_{i_1i_2}x_{i_3i_4}...x_{i_{n-1}i_n} 
$$
特定の$i_3, i_4, ..., i_n$に対し、$x_{12}$を含む場合は2p通り、$x_{1i_3}x_{2_i4}$を含む場合は2p 2(p-1) = 4p(p-1)通りあるので、
$x_{12}$を含む項の係数が$2p\frac{1}{2^pp!} = \frac{1}{2^{p-1}(p-1)!}$、$x_{1i_3}x_{2_i4}$の係数が、$4p(p-1)\frac{1}{2^pp!} = \frac{1}{2^{p-2}(p-2)!}$となり、一致する。
