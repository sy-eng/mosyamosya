そもそもコーシーの行列式がどうなっているかを考える。
$$
\begin{eqnarray}
C :=
\begin{vmatrix}
\frac{1}{x_i - y_j}
\end{vmatrix}_{i, j = 1, \cdots, n}
=
\begin{vmatrix}
\frac{1}{x_1 - y_1} & \cdots & \frac{1}{x_1 - y_n}\\
\vdots & & \vdots\\
\frac{1}{x_n - y_1} & \cdots & \frac{1}{x_n - y_n}\\
\end{vmatrix}
\end{eqnarray}
$$
問題に書かれているように、
$$
\begin{eqnarray}
C =
\begin{vmatrix}
\frac{1}{x_1 - y_1} & \cdots & \frac{1}{x_1 - y_n}\\
\vdots & & \vdots\\
\frac{1}{x_n - y_1} & \cdots & \frac{1}{x_n - y_n}\\
\end{vmatrix}
=\frac{\prod_{i < j} (x_i - x_j)(y_j - y_i)}{\prod_{i, j = 1}^{n} (x_i - y_j)}
\end{eqnarray}
$$
を示す。  
証明する前に$n = 3$で成立していることを確かめてみる。(n = 2は平鍋さんが書かれている)  
$$
\begin{eqnarray}
fC_{n=3} = \prod_{i, j = 1}^{3}(x_i - y_j)
\begin{vmatrix}
\frac{1}{x_1 - y_1} & \frac{1}{x_1 - y_2} & \frac{1}{x_1 - y_3}\\
\frac{1}{x_2 - y_1} & \frac{1}{x_2 - y_2} & \frac{1}{x_2 - y_3}\\
\frac{1}{x_3 - y_1} & \frac{1}{x_3 - y_2} & \frac{1}{x_3 - y_3}\\
\end{vmatrix}
=
\begin{vmatrix}
(x_2 - y_1)(x_3 - y_1) & (x_2 - y_2)(x_3 - y_2) & (x_2 - y_3)(x_3 - y_3)\\
(x_1 - y_1)(x_3 - y_1) & (x_1 - y_2)(x_3 - y_2) & (x_1 - y_3)(x_3 - y_3)\\
(x_1 - y_1)(x_2 - y_1) & (x_1 - y_2)(x_2 - y_2) & (x_1 - y_3)(x_2 - y_3)\\
\end{vmatrix}
=(x_2 - y_1)(x_3 - y_1)((x_1 - y_2)(x_3 - y_2)(x_1 - y_3)(x_2 - y_3) - (x_1 - y_3)(x_3 - y_3)(x_1 - y_2)(x_2 - y_2))
-(x_1 - y_1)(x_3 - y_1)((x_2 - y_2)(x_3 - y_2)(x_1 - y_3)(x_2 - y_3) - (x_2 - y_3)(x_3 - y_3)(x_1 - y_2)(x_2 - y_2))
+(x_1 - y_1)(x_2 - y_1)((x_2 - y_2)(x_3 - y_2)(x_1 - y_3)(x_3 - y_3) - (x_2 - y_3)(x_3 - y_3)(x_1 - y_2)(x_3 - y_2))
=(x_2 - y_1)(x_3 - y_1)(x_1 - y_2)(x_1 - y_3)((x_3 - y_2)(x_2 - y_3) - (x_3 - y_3)(x_2 - y_2))
-(x_1 - y_1)(x_3 - y_1)(x_2 - y_2)(x_2 - y_3)((x_3 - y_2)(x_1 - y_3) - (x_3 - y_3)(x_1 - y_2))
+(x_1 - y_1)(x_2 - y_1)(x_3 - y_2)(x_3 - y_3)((x_2 - y_2)(x_1 - y_3) - (x_2 - y_3)(x_1 - y_2))
=(x_2 - y_1)(x_3 - y_1)(x_1 - y_2)(x_1 - y_3)(-x_3y_3 - x_2y_2 + x_3y_2 + x_2y_3)
-(x_1 - y_1)(x_3 - y_1)(x_2 - y_2)(x_2 - y_3)(-x_3y_3 - x_1y_2 + x_3y_2 + x_1y_3)
+(x_1 - y_1)(x_2 - y_1)(x_3 - y_2)(x_3 - y_3)(-x_2y_3 - x_1y_2 + x_2y_2 + x_1y_3)
=(x_2 - y_1)(x_3 - y_1)(x_1 - y_2)(x_1 - y_3)(x_2 - x_3)(y_3 - y_2)
-(x_1 - y_1)(x_3 - y_1)(x_2 - y_2)(x_2 - y_3)(x_1 - x_3)(y_3 - y_2)
+(x_1 - y_1)(x_2 - y_1)(x_3 - y_2)(x_3 - y_3)(x_1 - x_2)(y_3 - y_2)
=\cdots=(x_1 - x_2)(x_1 - x_3)(x_2 - x_3)(y_3 - y_2)(y_3 - y_1)(y_2 - y_1)
\end{eqnarray}
$$
となり成立している。(といいつつ一部計算していない。。。)  
(i)について、検討する。fCを考える。教科書P53の(12)の右辺からの左辺への変形を繰り返し適用し、  
$$
\begin{eqnarray}
fC =
\begin{vmatrix}
\frac{\prod_{i = 1}^{n} (x_i - y_1)}{x_1 - y_1} & \cdots & \frac{\prod_{i = 1}^{n} (x_i - y_n)}{x_1 - y_n}\\
\vdots & & \vdots\\
\frac{\prod_{i = 1}^{n} (x_i - y_1)}{x_n - y_1} & \cdots & \frac{\prod_{i = 1}^{n} (x_i - y_n)}{x_n - y_n}\\
\end{vmatrix}
=
\begin{vmatrix}
\prod_{i = 2}^{n} (x_i - y_1) & \cdots & \prod_{i = 2}^{n} (x_i - y_n)\\
\vdots & & \vdots\\
\prod_{i = 1}^{n - 1} (x_i - y_1) & \cdots & \prod_{i = 1}^{n - 1} (x_i - y_n)\\
\end{vmatrix}
\end{eqnarray}
$$
上記の式より、分母がなくなるので、単項式もしくは多項式となる。(所謂、有理式でない。(単項式も多項式も有理式ではある。))  
$i, j$列$(i \neq j)$を比較すると,$y_i = y_j$のとき、$i, j$列の値がおなじになるのでP54の(17)より、行列式が0になる。  
つまり、$fC$は$(y_i-y_j)$で割り切れる。(この部分は(ii)になっている。)よって、単項式でなく、多項式になっている。  
(ii)について、考えてみる。まず、(i)の最後の付近より、$fC$は$(y_i-y_j)$にて割り切れる。  
これは、すべての$i \neq j$の組み合わせにて、言えるので、$fC$は$\prod_{i < j}(y_i-y_j)$にて割り切れる。  
上記を列でなく、行に対して、行うと$fC$は$\prod_{i < j}(x_i-x_j)$で割り切れる。
よって、$fC$は$\prod_{i < j}(x_i-x_j)(y_i-y_j)$にて割り切れる。  
P55の例3と同様に次数を比較してみる。  
問題のCの元になる行列は$x, y$に関して、-1次になっている。よって、行列式は$x, y$に関して、-n次。  
問題文の右辺の次数を考える。  
$\prod_{i < j}(x_i-x_j)$の次数を考えると相方の数だけになるので、1からn-1までの和$\frac{n(n-1)}{2}$次になる。  
これは$y$に関しても同様なので、分子は$\frac{n(n-1)}{2} + \frac{n(n-1)}{2} = n^2 - n$次になる。  
分母は$i, j$ともに1からnまであるので、$n^2$次になる。  
以上から、問題文の右辺は$n^2 - n - n^2 = -n$次になる。  
よって、問題の行列式と右辺の比較から、ある定数cで、
$$
\begin{eqnarray}
C =
\begin{vmatrix}
\frac{1}{x_1 - y_1} & \cdots & \frac{1}{x_1 - y_n}\\
\vdots & & \vdots\\
\frac{1}{x_n - y_1} & \cdots & \frac{1}{x_n - y_n}\\
\end{vmatrix}
=c\frac{\prod_{i < j} (x_i - x_j)(y_i - y_j)}{\prod_{i, j = 1}^{n} (x_i - y_j)}
\end{eqnarray}
$$
となる。(ここではcとCが異なっていることに注意。)
ここで、$fC$を再度考えて、cを求める。上の方の式から
$$
\begin{eqnarray}
fC = 
\begin{vmatrix}
\prod_{i = 2}^{n} (x_i - y_1) & \cdots & \prod_{i = 2}^{n} (x_i - y_n)\\
\vdots & & \vdots\\
\prod_{i = 1}^{n - 1} (x_i - y_1) & \cdots & \prod_{i = 1}^{n - 1} (x_i - y_n)\\
\end{vmatrix}
=c\prod_{i < j} (x_i - x_j)(y_i - y_j)
\end{eqnarray}
$$
$y_1$を考えると右辺の$y_i - y_j$について、$i = 1$にしないと$i < j$が成立しない。  
$\prod_{i = 1}^{n}{x_i}^{n-i}{y_i}^{n-i}$の項の係数を見ると、右辺の係数は$c$になる。  
左辺の行列式を見ると上記の項は今の行列式の対角成分の左上から$y_1^{n-1}, x_1y_2^{n-2}, \cdots, x_1x_2\cdots x_{n-1}$をかけたときしかない。  
(最初のn=3の例を参考のこと。x_i - y_jでi < jのときだけ、xをそうでないとき-yをかける。ここをわかりやすくするために上の例を  
n-2でなくn=3とし、fCを考慮した。)  
すると、左辺の係数は$-y$に影響されて、$(-1)^{\frac{n(n-1)}{2}}$となる。
(i)(ii)から、$n(n-1)$個ある、$y_i - y_j$に-1をかけて、問題にある式を得る。
