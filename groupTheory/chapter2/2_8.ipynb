## (4)
まず、Hが部分群であることを確認する。(ここまでの問題に関して、部分群であることはこれまでに確認されている。)
命題2.3.2に従って確認する。
少なくとも部分集合であることは良い。

(1)に関して、Hに1が含まれているので良い。

(3)に関して、Hのそれぞれの元は、自分自身が逆元になっている。

(2)に関して、演算は$4 \times 4$=16通りあるが、そのうち、1とそれ以外の演算、6通りはHに含まれることは明らか。さらに、同じものをかける4通りも、逆元になるので、1になり、これもHに含まれる。

つまり、残りの6通りを考える。
$$
(1, 2)(3, 4)(1, 3)(2, 4) = (1, 2)(3, 4)((1, 2)(2, 3)(1, 2)(2, 3)(3, 4)(2, 3))
=(3, 4)(2, 3)(1, 2)(2, 3)(3, 4)(2, 3)=(3,4)(1,3)(3,4)(2,3)=(1,4)(2,3)\\
(1, 3)(2, 4)(1, 2)(3, 4) = ((2, 3)(3, 4)(2, 3)(1, 2)(2, 3)(1, 2))(1, 2)(3, 4)
=(2, 3)(3, 4)(2, 3)(1, 2)(2, 3)(3, 4)=(2, 3)(3, 4)(1, 3)(3, 4)=(2, 3)(1, 4)=(1, 4)(2,3)\\
$$
同様に、
$$
(1, 2)(3, 4)(1, 4)(2, 3) = (1, 3)(2, 4)\\
(1, 4)(2, 3)(1, 2)(3, 4) = (1, 3)(2, 4)\\
(1, 3)(2, 4)(1, 4)(2, 3) = (1, 2)(3, 4)\\
(1, 4)(2, 3)(1, 3)(2, 4) = (1, 2)(3, 4)\\
$$
よって、(2)が満たされ、部分群になる。

さて、$ghg^{-1}$を考える。

gが互換の積ABC(実際は最大3個の積で表されるが、もっと多くても、少なくても問題ない。)で表されるとする。
その時、$g^{-1}$はCBAと表される。もちろん恒等置換であれば、hはhになり、Hに含まれる。

このとき、ChCを考える。hは(a,b)(c,d)のように表され、任意の組み合わせが考えられる。このとき、Cが(a,b)もしくは(c,d)のときは
Cとhが可換になり、ChC=CCh=hとなり、Hに含まれる。

Cは互換だが、$_4C_2=6$通りある。2通りは確認できたので、残り4通りを確認する。

Cが(a,d)のとき、
$$
ChC = (a,d)(a,b)(c,d)(a,d) = (a,c)(c,d)(a,c)(c,d)(a,b)(a,d) = (a,c)(a,d)(a,b)(a,d) = (a,c)(b,d)
$$
となり、Hに含まれる。(a,b)=(b,a),(c,d)=(d,c)であることを考えると、他の互換でも、Hに含まれる。
$$
(a,c)(a,b)(c,d)(a,c) = (a,d)(b,c)\\
(b,c)(a,b)(c,d)(b,c) = (a,c)(b,d)\\
(b,d)(a,b)(c,d)(b,d) = (a,d)(b,c)\\
$$
更にA,Bを適用させるときも、Hに含まれる。よって、$ghg^{-1} \in H$となる。よって、正規部分群。

# 2.8.2
Hは群Gの指数2の部分群になっている。このとき、定義2.6.7にある同値類に関して考えてみる。

まず、$h_1, h_2 \in H$に関して考えると、Hは部分群なので$h_1 h_2, h_2 h_1^{-1} \in H$(これらが等しいとは限らない)であり、右剰余類としても、$C(h_1) = H$となる。(左剰余類として、$c \notin H$で$c \in C(h_1)$であるとする。すると、$h_1^{-1}c = h_3 \in H$となるが、c = h_1 h_3となり、右辺はHに含まれ、矛盾する。右剰余類だとしても同様。)

(G:H)=2なので、剰余類は2個しかなく、Hがそのうちのひとつなので、剰余類は(右でも、左でも)HとCとすることができ、
$G = H \cup C, H \cap C = \empty$となる。(定理2.6.20からも|G| = (G:H)|H| = 2|H|)
つまり、右剰余類としても、左剰余類としても、Gに対して、指数2の部分群が決まると、Cが決まる。

ここで、任意の$h \in H, c \in C$を考える。

まず、hを考える。
すべての$h_1 \in H$に関して、$h h_1 h^{-1} \in H$なので、正規部分群の条件を満たしている。

次にcを考える。
すべての$h_1 \in H$に関して、$c h_1 c^{-1}$はHに含まれるか、Cに含まれる。

もし、Hに含まれる場合、ある$h_2$が存在するが、$c h_1 c^{-1} = h_2$で、$c h_1 = h_2 c$となり、
右剰余類でも、左剰余類でも同じCになるため、$c h_1 = h_2 c = c_1 \in C$となりうる。

もし、Cに含まれる場合、ある$c_2$が存在し、$c h_1 c^{-1} = c_2, c = c_2 h_3 = h_4 c_2 (c_2 \in C)$となる。
すると、$c_2 h_3 h_1 (c_2 h_3)^{-1} = c_2 h_3 h_1 h_3^{-1} c_2^{-1} = c_2$となり、
$c_2 = (h_3 h_1 h_3^{-1})^{-1} \in H$となり、矛盾する。よって、cに関しても正規部分群の条件を満たす。

よって、Hは正規部分群になる。

# 2.8.3
まず、$N_1 N_2$がGの部分群であることを確認して、定義2.8.1に従い、$N_1 N_2$が正規部分群であることを確認する。

$N_1 N_2$がGの部分群であることを確認する。これを確認するには命題2.3.2を利用する。

(1)について、$N_1,N_2$はGの部分群であるため、命題2.3.2より$1_G = 1_{N_1} \in N_1, 1_G = 1_{N_2} \in N_2$となる。
$N_1 N_2$はそれぞれの集合(部分群)の要素の積の写像になるので、$1_G = 1_{N_1} 1_{N_2} \in N_1 N_2$となり、(1)を満たす。

(2)について、考える。
${n_1}_1, {n_1}_2 \in N_1, {n_2}_1, {n_2}_2 \in N_2$として、$N_1 N_2$の元が、
${n_1}_1 {n_2}_1$及び、${n_1}_2 {n_2}_2$とできる。
このときに、${n_1}_1 {n_2}_1 {n_1}_2 {n_2}_2 \in N_1 N_2$であることを言えると良い。

ここで、補題2.8.10を利用する。${n_2}_1 \in N_2 \subset G$であるため、以下のようになる。
${n_2}_1 N_1 = N_1 {n_2}_1$。つまり、${n_2}_1 {n_1}_2 = {n_1}_3 {n_2}_1$となる、${n_1}_3 \in N_1$が存在する。すると、${n_1}_1 {n_2}_1 {n_1}_2 {n_2}_2 = {n_1}_1 {n_1}_3 {n_2}_1 {n_2}_2$となるが、$N_1, N_2$が部分群なので、命題2.3.2の(2)より、${n_1}_1 {n_1}_3 = {n_1}_4 \in N_1$, ${n_2}_1 {n_2}_2 = {n_2}_3 \in N_2$となり、${n_1}_1 {n_2}_1 {n_1}_2 {n_2}_2 = {n_1}_1 {n_1}_3 {n_2}_1 {n_2}_2 = {n_1}_4 {n_2}_3 \in N_1 N_2$となり、(2)を満たす。

(3)について、考える。流れは(2)とおなじになる。
${n_1}_1 \in N_1, {n_2}_1 \in N_2$として、$N_1 N_2$の元が、${n_1}_1 {n_2}_1$とできる。
このときに、$({n_1}_1 {n_2}_1)^{-1} = ({n_2}_1)^{-1} ({n_2}_1)^{-1} \in N_1 N_2$であることを言えると良い。

ここで、命題2.3.2、補題2.8.10を利用する。$({n_2}_1)^{-1} \in N_2 \subset G$であるため、以下のようになる。
$({n_2}_1)^{-1} N_1 = N_1 ({n_2}_1)^{-1}$。つまり、$({n_2}_1)^{-1} ({n_1}_1)^{-1} = {n_1}_2 ({n_2}_1)^{-1}$となる、${n_1}_2 \in N_1$が存在する。すると、$({n_1}_1 {n_2}_1)^{-1} = ({n_2}_1)^{-1} ({n_2}_1)^{-1} = {n_1}_2 ({n_2}_1)^{-1} \in N_1 N_2$となり、(3)を満たす。

上記から、命題2.3.2の(1)-(3)を満たすことがわかったので、$N_1 N_2$はGの部分群になっていることがわかった。

次に、これが、定義2.8.1を満たし、正規部分群になることを確認する。
そもそも、$N_1, N_2$が正規部分群なので、すべてのgに対して、すべての${n_1}_1 \in N_1, {n_2}_1 \in N_2$に対して、ある、${n_1}_2 \in N_1, {n_2}_2 \in N_2$があり、${n_1}_2 = g {n_1}_1 g^{-1}, {n_2}_2 = g {n_2}_2 g^{-1}$となる。($g = g^{-1} \in G$を考えることで、${n_1}_2, {n_2}_2$に関しても、例えば、${n_1}_1 =  g^{-1} {n_1}_2 g$となる。)

${n_1}_2 {n_2}_2 \in N_1 N_2$に対して、共通のgを考え、${n_1}_2 {n_2}_2 = g {n_1}_1 g^{-1} g {n_2}_1 g^{-1} = g {n_1}_1 {n_2}_1 g^{-1}$となる。${n_1}_1 {n_2}_1 \in N_1 N_2$なので、すべての${n_1}_1 {n_2}_1$に対して、すべてのgで$N_1 N_2$に含まれる。

よって、正規部分群になることがわかった。

# 2.8.4

まず、$\mathfrak{S}_3$の部分群を考える。

演習2.1.3を参考にすると、$|\mathfrak{S}_3| = 6$となる。

例2.3.4から$1,\mathfrak{S}_3$は部分群となる。
これらの位数は$|1| = 1, |\mathfrak{S}_3| = 6$となる。

命題2.3.13より、まず、ある元が生成する部分群を考える。
$\mathfrak{S}_3$の元の位数は1(1),2((1,2), (2,3), (1,3)),3((1,2,3), (1,3,2))になる。
位数2に関して、$|<(i,j)>|=2$,位数3に関して、$|<(i,j,k)>|=3$になり、位数3の場合は位数3の元と、1で群になっている。

系2.6.21より、部分群の位数は1,2,3,6しかありえない。
位数1の部分群は部分群が1を含むので\{1\}しかあり得ない。また、6の部分群も$\mathfrak{S}_3$になる。

2のときは、1とその他の元になるが、位数2の元以外にはならないので上記の3種類の部分群$<(i,j)>,(i \neq j)$になる。

部分群で位数3のときは上記に位数3の元2個と、1で成り立つときがあるが、それ以外の場合がないことを確認する。

1は必ず含まなければならず、位数3の元があると、残りの位数3の元も含まれてしまうので、ありうるとしたら、位数2の元2個と、1となる時になる。
しかし、特に$\mathfrak{S}_3$のときは、演習2.3.9を踏まえると、どの組み合わせでも全体を生成してしまう。そのため、位数2の部分群にはならない。

よって、$\mathfrak{S}_3$の部分群は以下の6通りになる。
$1, \mathfrak{S}_3, <(1,2)>, <(2, 3)>, <(1, 3)>, <(1, 2, 3)>$

次にこの中で、正規部分群になるものを考える。

定義2.8.1を考えると、1は明らかに満たすし、$\mathfrak{S}_3$も群の性質から定義を満たす。そのため、これらは正規部分群になる。
また、例2.8.9より<(1,2,3)>も正規部分群になる。

残りの3個に対して、正規部分群になるかを確認する。
$h=(i, j) \in <(i,j)>=\{1, (i, j) \}=H$として、$g = g^{-1} = (j, k)$とする。このとき、
$ghg^{-1} = (j, k)(i, j)(j, k) = (i, k)$となり、これはHに含まれない。そのため、これらは正規部分群にならない。

そのため、$\mathfrak{S}_3$の正規部分群は$1, \mathfrak{S}_3, <(1, 2, 3)>$になる。


# 2.8.5
そもそも、四元数群は例2.3.11に記載がある。

まず、これの部分群を考える。

四元数群Hは位数8になる。

例2.3.4から$1,H$は部分群となる。
これらの位数は$|1| = 1, |H| = 2$となる。

系2.6.21より、部分群の位数は1,2,4,8しかありえない。
位数1の部分群は部分群が1を含むので\{1\}しかあり得ない。また、8の部分群も$\mathfrak{S}_3$になる。

命題2.3.13より、まず、ある元が生成する部分群を考える。
Hの元の位数は1(1),2 (-1), 4 $(\pm i, \pm j, \pm k)$になる。このとき部分群は$\{1, a, -1, -a \}(a \in {i, j, k})$になる。これは3種類だけ存在する。

位数が2のときは、1とその他の元になるが、位数2の元以外にはならないので上記の1種類の部分群になる。

部分群で位数4のときは上記3種類以外の場合がないことを確認する。

上記で1つの生成元から生成される部分群を考えたので、そうならない4つの元からなる、部分群があるか考える。
位数4でないものは2個しかないので、位数4の部分群には位数4の元が含まれる。ある位数4の元、aが含まれると、群になるには、1,-1,-aが含まれる。
この時点で、他の元bが含まれる、余地がないので、上記の3種類しかない。

よって、部分群は、\{1\}, H, <-1>, < i >, < j >, < k >の6種類になる。

さて、部分群の中で、正規部分群になるものを考える。
定義2.8.1を考慮すると、\{1\}, H, <-1>は正規部分群になる。

次に< i >, < j >, < k >を命題2.8.7に当てはめる。

Hの生成元を考える。i,jもしくはj,kもしくはi,kを生成元とし、それをa,b、それ以外をcとすると、
aが生成元なら、1,-1,a,-aが含まれ、その上でbがあるので、b,-bも含まれる。abがcになり、その積(c^3)で、-cになるので、
cも生成される。そのため、上記の組み合わせは生成元となることがわかる。

さて、上記の記号で部分群が< a >の時を考える。$S=\{ a, b\}, T = \{ a \}$になり、$a a a^{-1} = a \in <a>, b a b^{-1} = - b a b = -b c = -a \in <a>$
(この式でab = -baになる。a,bの選択によってはab = -cになりうるが、符号の違いのみなので、$b a b^{-1} \in <a>$という結論は変わらない。)
よって、命題2.8.7より、< i >, < j >, < k >も正規部分群であることが確認できた。

よって、Hのすべての部分群が正規部分群となる。
