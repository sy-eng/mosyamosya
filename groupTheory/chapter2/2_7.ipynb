# 2.7.1

Gの元、恒等写像$1_G$と(1,4)が完全代表系になる。Hは1から3の置換になるため、$h_1,h_2, h_3, h_4 \in H$と、$g_1 = 1_G, g_2 = (1, 4) \in G$に関して、$g_1,g_2$が同値類だとすると$h_1 g_1 h_2 = h_3 g_2 h_4$となる。これを計算すると、$h_3^{-1} h_1 h_3 h_4^{-1} = (1, 4)$となる。
左辺はhの積なので部分群Hに含まれるが、右辺はHに含まれず、矛盾している。よって、$g_1,g_2$は同値類でない。よって、$1_G$と(1,4)は完全代表系になりうる。

一方で、この2つの元の同値類で、Gが表されることを示す。
任意の$g_1 \in G$が、$h_1,h_2 \in H$と、$g_2 \in G$で、$g_1 = h_1 g_2 h_2$となることを確認する。
$g_1$で4が4のままのものと、4以外のiになるものを考える。
まず、4のままのものを考えると、$g_1$はHの元になる。それを$h_1$とすると、$h_2 = 1_H = 1_G$とすると、$g_1 = h_1 g_2 h_2$となる。
よって、任意の4が置換されない、$g_1 \in G$が、$h_1,h_2 \in H$と、$g_2 \in G$で、$g_1 = h_1 g_2 h_2$となる。

次に、4が4でない、iとなるとする。このとき、$h_2=(1,i)$、$g_2 = (1, 4)$とすると、4はiに置換され、$h_1$は自由に1から3番目の要素を
入れ替えられるので、任意の4が4以外のiに置換される、$g_1 \in G$が、$h_1,h_2 \in H$と、$g_2 \in G$で、$g_1 = h_1 g_2 h_2$となる。

よって、恒等写像$1_G$と(1,4)が完全代表系になる。

## (1)
以下のように設定する。
$$
\begin{equation}
b1 = 
\begin{pmatrix}
b_{111}  \\
\vdots & \ddots & \Huge{0} \\
b_{1n1} & \cdots & b_{1nn}\
\end{pmatrix}
\end{equation}
$$
$$
\begin{equation}
b2 = 
\begin{pmatrix}
b_{211}  \\
\vdots & \ddots & \Huge{0} \\
b_{2n1} & \cdots & b_{2nn}\
\end{pmatrix}
\end{equation}
$$
$$
\begin{equation}
g = 
\begin{pmatrix}
g_{11} & \cdots & g_{1n} \\
\vdots & \ddots & \vdots \\
g_{n1} & \cdots & g_{nn}\
\end{pmatrix}
\end{equation}
$$
このとき、
$$
h =
\begin{pmatrix}
\sum_{q=1}^{n} b_{2q1} b_{111} g_{1q} & \cdots & \sum_{q=i}^{n} b_{2q1} b_{111} g_{1q} & \cdots & b_{2nn} b_{111} g_{1n} \\
\vdots & & \vdots & & \vdots \\
\sum_{q=1}^n b_{2q1} \sum_{p=1}^{i} b_{1ip}g_{pq} & \cdots & \sum_{q=j}^n b_{2qj} \sum_{p=1}^{i} b_{1ip}g_{pq} & \cdots & b_{2nn} \sum_{p=1}^{i} b_{1ip}g_{pn}\\
\vdots & & \vdots & & \vdots \\
\sum_{q=1}^n b_{2q1} \sum_{p=1}^{n} b_{1np}g_{pq} & \cdots & \sum_{q=j}^n b_{2qj} \sum_{p=1}^{n} b_{1np}g_{pq} & \cdots & b_{2nn} \sum_{p=1}^{n} b_{1np}g_{pn}\\
\end{pmatrix}
$$
つまり、
$$
h_{ij} = \sum_{q=j}^n b_{2qj} \sum_{p=1}^{i} b_{1ip}g_{pq}
$$
$h_{i n} = b_{2nn} \sum_{p=1}^{i} b_{1 i p}g_{pn}$を考える。

$b_1, b_2 \in B \subset G = GL_n(\mathbb{R})$なので、$b_1, b_2$は正則で$b_{ii} \neq 0$となる。

$i \leq i_n$に対して、$b_{1ij} = 0 (i \neq j), b_{1ii} = 1$とすると、$h_{i n} = 0 (i \neq j), h_{i n} = g_{i_n n} \neq 0$となる。
$i > i_n$に対して、$b_{1ij} = 0 (i, i_n \neq j), b_{1ii} = 1, b_{1 i i_n} = - \frac{g_{i n}}{g_{i_n n}}$とすれば、$h_{i n} = 0$とでき、
全体として、$b_1$の対角成分が0でなくなり、正則になる。

$h_{i_n j} = \sum_{q=j}^n b_{2qj} \sum_{p=1}^{i_n} b_{1 i_n p}g_{pq}$を考える。
上記の$b_1$を考慮すると、$h_{i_n j} = \sum_{q=j}^n b_{2qj} b_{1 i_n i_n}g_{i_n q} = \sum_{q=j}^n b_{2qj} g_{i_n q}$

$b_2$が正則になるので、$b_{2jj} \neq 0$になる。$j \neq n$のとき、上記の式で、0であってはいけない変数は$g_{i_n n}, b_{j j}$になる。
$b_{2 i j}$に関しては自由に設定できるが、$g_{i j}$に関しては、与えられる。

よって、ひとまず、$b_{2 i i} = 1$として、$b_{2 i j} = 0(j \neq i, j \neq n)$とすると、
$h_{i_n j} = \sum_{q=j}^n b_{2qj} g_{i_n q} = g_{i_n j} + b_{2 j n} g_{i_n n}$となる。
これが0になるようにすると、$b_{2 j n} = -\frac{g_{i_n j}}{g_{i_n n}}$とすればよい。

上記から、正則な下三角行列$b_1, b_2$が存在する。
