# 演習2.3.2

命題2.3.2を使って、部分群であることを示す。

定義よりHの要素gは$g \in G = GL_{2n}(\mathbb{R})$となり、HはGの部分集合となる。
しかし、gを2n次の正方行列としても、$det(J_n)=1$となり(行列をn回入れ替えて、対角行列とできる。かつ、-1がn個ある。)、$det(g^T J_n g) = det(g^T)det(J_n)det(g) = det(g)^2 = 1 \neq 0$から、$det(g) \neq 0$となり、gは正則になる。

(1)が成立することを示す。
${I_G}^T J_n I_G = I_G J_n I_G = J_n$となるため、$I_G \in H$で(1)が成立する。

(2)が成立することを示す。
任意の$x,y \in H$に対して、$(xy)^T J_n (xy)$を考える。

$(xy)^T J_n (xy) = y^T x^T J_n x y = y^T (x^T J_n x) y = y^T J_n y = J_n$

よって、$xy \in H$になる。

(3)が成立することを示す。
任意の$x \in H$に対して、${x^{-1}}^T J_n x^{-1}$を考える。

${x^{-1}}^T J_n x^{-1} = {x^{-1}}^T x^T J_n x x^{-1} = J_n$

よって、$x^{-1} \in H$になる。

(1)-(3)が成立するので、Hは$GL_{2n}(\mathbb{R})$の部分群。

# 演習2.3.3

命題2.3.2を使って、部分群であることを示す。

定義よりHの要素gは$g \in G = GL_{n}(\mathbb{C})$となり、HはGの部分集合となる。
しかし、gをn次の正方行列としても、$\overline{g^T}g = I_n$より、$det(\bar{g^T}g) = \overline{det(g^T)}det(g) = \overline{det(g)}det(g) = |det(g)|^2 = det(I_n) = 1$
よって、gは正則になる。

命題2.3.2の(1)が成立することを示す。
$g = I_n \in G$は$\overline{g^T} = I_n$となるため、$\overline{g^T}g = I_n$となり、$g = I_n \in H$となる。

(2)が成立することを示す。
任意の$x,y \in H$に対して、$\overline{(xy)^T} (xy)$を考える。
$\overline{(xy)^T} (xy) = \overline{y^T}\overline{x^T}xy = \overline{y^T} I_n y = \overline{y^T} y = I_n$
よって、$xy \in H$

(3)が成立することを示す。
任意の$x \in H$に対して、$\overline{(x^{-1})^T} x^{-1}$を考える。
$\overline{(x^{-1})^T} x^{-1} = \overline{(x^T)^{-1}} x^{-1} = \overline{(x^T)^{-1}} I_n x^{-1} = \overline{(x^T)^{-1}} \overline{x^T} x x^{-1} = \overline{I_n}I_n = I_n I_n = I_n$
よって、$x^{-1} \in H$になる。

(1)-(3)が成立するので、Hは$GL_{n}(\mathbb{C})$の部分群。

# 演習2.3.7
一般に複素数は以下のように表される。
$$
z = r \, e^{i\theta} = r(cos \theta + i \, sin \theta) \qquad (r \in \mathbb{R}, r \geq 0, 0 \leq \theta < 2\pi)
$$
このとき、$z^n$は加法定理を考慮すると、$z = r^n \, (cos \, n \theta + i \, sin \, n \theta)$となる。

これが、$1 = 1 (cos (2m\pi) + i \, sin (2m\pi)) (m \in \mathbb{Z})$となることから、$r^n = 1$, $2m\pi = n \theta$となる。

$r^n$は正の実数の範囲で0以上の値域を取り、単調増加の関数になるため、$r^n = 1$は唯一の解を持ち、$r = 1$になる。

また、$\theta = \frac{2m\pi}{n} (m \in \mathbb{Z}, 0 \leq m < n)$となり、解がn種類あることがわかる。

先の加法定理を考慮すると、$H=<e^{\frac{2 \pi i}{n}}>$となる。($H=<e^{\frac{-2 \pi i}{n}}>$も成立する。)

また、$(e^{\frac{2 \pi i}{n}})^n = 1$になり、n以下では1にはならないため、位数はnになる。

なお、$m \neq 1$の場合は、n回かける前に1になる場合がありうる。(ならない場合もある。当然、n回かけると1にはなる。)
そのため、そのような元では、n個の元をすべて、生成することはできない。
よって、任意のmが位数nの生成元になるとは言えない。

# 演習2.3.9
任意の置換が、互換の積で表されることを示す。

帰納法で示す。

n=1, 2のときは明らかに成立する。n=1は入れ替えるものがないので、恒等置換しかないので成立する。
n=2のときは恒等置換と、互換しかないので、成立する。

n-1個のときに任意の置換が互換の積で表されるとする。

n個のときにn番目の置換先がnのときはnが変更されないので、実質、n-1個の置換になる。

n番目の置換先がjで置換後、n番目にkが来るとき、n番目の置換先がn、j番目がkになる置換は上記の実質n-1個の置換で表される。
これを$\sigma$とすると、この置換は$(n, j)\sigma$となり、(n,j)が互換で、$\sigma$も互換の積なので、全体も互換の積で表される。

よって、各問題では、与えられた置換で、任意の互換が表されれば、示された元で、$\mathfrak{S}_n$が生成されることがわかる。
## (1)
任意の互換が$\sigma_1, ..., \sigma_{n-1}$の積で表されることを示す。

(j, k)(j < k)は$\sigma_{j}\sigma_{j+1}...\sigma_{k-2}\sigma_{k-1}\sigma_{k-2}...\sigma_{j+1}\sigma_j$で表される。
j > kだとしても、互換はおなじになるので、j < kだけを考えれば良い。
(直感的にはjを一つずつ、後ろに移動させ、その後、kを一つずつ前に持ってくる。)

なお、ある$\sigma_i$がない場合、ある、$a(a \leq i)$と$b(b \geq i+1)$を入れ替えることができなくなる。つまり$\mathfrak{S}_n$の
元である互換(a, b)が表されなくなり、$\mathfrak{S}_n$が生成されなくなる。
## (2)
任意の互換が$\sigma, \tau$の積で表されることを示す。

これを示すために、(1)の任意の$\sigma_i$が$\sigma, \tau$の積で表されることを示す。

$\sigma_i = \sigma^{n-i+1}\tau\sigma^{i - 1}$となるので、任意の$\sigma_i$は$\sigma, \tau$の積で表される。
(直感的にはi, i+1が1,2番目に来るようにぐるぐる回して、i, i+1を入れ替える。その後、もとに戻るようにぐるぐる回す。)

よって、$\sigma, \tau$で生成される。

なお、$\tau^2$は1となる。そのため、$\tau$だけの積で$\sigma$を生成することはできない。
また、
$$
\tau
=
\left(\begin{array}{ccccc}
1 & 2 & 3 & \cdots & n \\
2 & 1 & 3 & \cdots & n \\
\end{array}
\right)
$$
であるが、
$$
\sigma^i
=
\left(\begin{array}{ccccc}
1 & 2 & 3 & \cdots & n \\
i+1 & i+2 & i+3 & \cdots & i \\
\end{array}
\right)
$$
となり、$\sigma^i \neq \tau$。よって、$\sigma, \tau$ともにないと、全体を生成できない。
