# 演習2.6.1
定義1.3.2で関係が定義されている。
このとき、例えば、x=1とすると、$R = \{(1, 1), (1, 2), (2, 1) \}$が関係があることになる。

これが定義2.6.1の(1)から(3)を満たす必要がある。

a = 1, b = 2, c = 4とすると、x=1としたときに(a,b)=(1,2)は関係がある。同様にx=2としたときに(b,c)=(2,4)は関係がある。
すると、(3)より、(a,c)も関係がないといけないが、(1,4)となる、xは存在しない。
よって、Rは同値関係でない。

なお、(1)については、(x,x)が含まれるため、成り立つ。

(2)について、x = yであれば、(1)も成り立つので、成り立つ。$x \neq y$であれば、x = 2y もしくは、2x = yとなる。
x=2yのとき、$x \sim y$が(x, y) = (2y, y)となるが、$y \sim x$が(y, x) = (y, 2y)となり、これはRに含まれる。
2x=yとしても、同様。そのため、(2)も成り立つ。

# 演習2.6.2
定義2.5.18(2)より、$h_1 = g h_2 g^{-1}$のとき、共役となる。

(1)について、考える。

任意のhに関して、$h = g h g^{-1}$となるgが存在する必要がある。Gは群なので、必ず単位元$I_g$が存在する。
すると、$g = I_g$とすると、$I_g h {I_g}^{-1} = h$となり、成立する。

(2)について、考える。

ある、$h_1, h_2 \in G$に関して、共役であるつまり、$h_1 = g h_2 g^{-1}$となる$g \in G$が存在するとする。
その時、$k = g^{-1}$とすると、$h_2 = k h_1 k^{-1} = g^{-1} g h_2 g^{-1} (g^{-1})^{-1}$となり、(2)も成立する。

(3)について、考える。

ある、$h_1, h_2, h_3 \in G$に関して、$h_1 = g_1 h_2 {g_1}^{-1}$となる$g_1 \in G$が存在し、かつ、
$h_2 = g_2 h_3 {g_2}^{-1}$となる$g_2 \in G$が存在するとき、$h_1 = g h_3 g^{-1}$となる$g \in G$となる、
gが存在することを示せば良い。
最初の2式を考えると、$h_1 = g_1 h_2 {g_1}^{-1} = g_1 (g_2 h_3 {g_2}^{-1}) {g_1}^{-1}$となる。
このとき、$g = g_1 g_2$とすると、Gは群なので、これはGの要素で、$h_1 = g h_3 g^{-1}$となる、gになりうる。
よって、(3)も満たす。

# 演習2.6.3
位数3の群をH,位数5の群をGとする。

定義2.1.2より、$|H| = 3, |G| = 5$となる。

背理法で考える、HがGの部分群だとすると、系2.6.21(1)より、|H|は|G|の約数になる。
しかし、3は5の約数でないので、矛盾している。

よって、HはGの部分群でなく、位数3の群は、位数5の群の部分群でないと言える。

# 演習2.6.4
H,KがGの有限部分群という仮定があるので、$1_G \in H$かつ、$1_G \in K$になっている。よって、$H \cap K \neq \emptyset$

また、命題2.3.3より、$H \cap K$はGの部分群になる。

命題2.3.3の証明は省略されているが、部分群をH,Kとして、以下のようになる。

集合を考え、$H \cap K \subset H \subset G, H \cap K \subset K \subset G$。

よって、命題2.3.2からGにおいて、(1)-(3)が、$H \cap K$で成り立つことを言えば、
$H \cap K$がGで部分群となることが言える。

(1)について、H,KはGの部分群なので、$1_G \in H \neq \emptyset, 1_G \in K \neq \emptyset$。
よって、$1_G \in H \cap K \neq \emptyset$。(空でないので、以下、元を選べる。)

(2)について、H,Kについての命題2.3.2(2)を利用すると、$a, b \in H \cap K \subset H$に対して、
$ab \in H \subset Gs$であり、かつ、これは、$a, b \in H \cap K \subset K$なので、$ab \in K \subset G$になるので、
$ab \in H \cap K$。

(3)について、H,Kについての命題2.3.2(3)を利用すると、$a \in H \cap K \subset H$に対して、
$a^{-1} \in H \subset G$であり、かつ、これは、$a \in H \cap K \subset K$なので、$a^{-1} \in K \subset G$になるので、
$a^{-1} \in H \cap K$。

よって、命題2.3.3が成り立っていることがわかる。

この命題では$H \cap K$がGの部分群になることを言っているが、上の証明をもとに、
H,Kで命題2.3.2を考えると、$H \cap K$はH,Kそれぞれで、部分群になっていることがわかる。

仮定より、H,Kそれぞれが、有限群なので、$H \cap K$も有限部分群となる。

系2.6.21より、$H \cap K$はH,Kそれぞれの部分群なので、$|H \cap K|$が$|H|,|K|$の公約数になる。

$|H|,|K|$が互いに素なので$|H \cap K| = 1$(公約数がこれ以外ない)になり、$H \cap K = \{1_G\}$
