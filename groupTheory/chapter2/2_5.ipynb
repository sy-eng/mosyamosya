# 演習2.5.1

そもそも、xを生成元とする、巡回群Gにおいて、元の個数がmだとすると、位数がmになる。

$x^s = 1$であるため、ある整数$a, r (0 \leq r < s)$を使って、$j = as + r$として、
$x^j = x^{as + r} = x^r$となり、元の数はたかだか位数sと同じ、s個しかないと言える。

もし、位数がmでなく、s>mだとする。その時、$x^s = 1$になるが、位数の定義より$j < s$に対して、$x^j \neq 1$ 。
一方で、元がm<sしかないので、$x, x^2, \cdots, x^{s-1}$のうち、いくつかは同じになっている。
そのとき、$x^{j_1}, x^{j_2}, (0 < j_1 < j_2 < s)$が同じだとする。
そうすると、命題2.3.19も踏まえると、$x^{j_2} x^{-j_1} = x^{j_2 - j_1} = 1$となる。
仮定より、$0 < j_2 - j_1 < s$となり、sが位数であることに反する。

上記より、巡回群Gにおいて、位数はmになることがわかる。また、、巡回群のm個の元は互いに異なる。
(これと同様の話は、どこかで証明していたような気もしないでないが、やっていない？)

## (1)
m,nの満たすべき必要十分条件がmがnの倍数だとする。つまり、ある整数$d \geq 0$で$m = nd$となる。

そもそも、$x^{i_1} = x^{i_2}$となるため、$x^{i_2 - i_1} = 1$となる。yに関しても同様のことが言える。

命題2.4.18より、すべての$x^{i_1} = x^{i_2}$となる、$i_1, i_2 \in \mathbb{Z}$に対して、
$i_2 - i_1 = c_x m, c_x \in \mathbb{Z}$となるが、$i_2 - i_1 = c_x m$となる、すべての$c_x$に対して、
$i_2 - i_1 = c_y n, c_y \in \mathbb{Z}$となる、$c_y$が存在するm,nの条件を求めることと同値になる。

まず、$m = nd$ならば、$i_2 - i_1 = c_x m$となっているので、$i_2 - i_1 = c_x m = c_x n d$なので、$c_y = c_x d$となり、
$c_y$が存在する。

次に、すべてのc_xに対して、$i_2 - i_1 = c_x m = c_y n$となる、$c_y$が存在するとする。このとき、$c_x = 1$でも成立するので、
$c_y = \frac{m}{n}$となり、mはnの倍数で$m = nd, d \in \mathbb{Z}$とならないといけない。$c_x$が1以外のときも$m = nd$であれば、
$c_y$は整数になる。

よって、mがnの倍数であることが必要十分条件になる。

## (2)
写像$\phi: G \rightarrow H$はHが群になる準同型の写像でないといけない。

mがnの倍数でない場合、well-definedでないことを確認する。well-definedの定義はP.10で行われているが、
今回は(2)が問題になる。これは、写像の定義の任意の元aに対し、Hの$\phi(a)$が一つに決まるというということが、
計算方法によって成立しないため起こる問題になる。つまり、ある元bに対して、対応するHの元が複数存在する。
計算方法が、複数あるのは$\phi$をGで計算したあとに適用することができ、かつ先に$\phi$を適用して、Hで計算することが
できるからである。

整数倍でないので、$m = dn + r, (d, r \in \mathbb{Z}, 0 < r < n)$とかける。

このとき、$\phi(x^{c_x m}) = \phi(x^{c_x(an + r)}) = \phi(1) \equiv y^0 = 1_H$となるが、一方で、
$\phi(1) = \phi(x^{c_x (an + r)}) \equiv y^{c_x (an + r)} = y^{c_x r}$。
すべての$c_x$で成立するので、1でも成立する。よって、$y^r = y^0$となり、rは0でなく、位数の定義から
$y^r$は1でないので、特に、$\phi(x^{m})$に関して、複数($1_H, y^r$)の対応する元を持ちうる。
よって、写像の定義として、成立しない。

一方で、整数倍のとき、$m = dn, (d \in \mathbb{Z})$となる。このとき、任意の$i = am + s, (a, s \in \mathbb{Z}, 0 \leq s < m)$となる。
$\phi(x^i) = \phi(x^{am + s}) = \phi(x^s) = y^s$となり、一方で、$\phi(x^{am + s}) = \phi(x^{adn + s}) = y^{adn + s} = y^s$となり、
特定の$\phi$の定義として、成立する。

もちろん、$\phi$は一般的に$\phi(x^{i_1}x^{i_2}) = \phi(x^{i_1 + i_2}) \equiv y^{i_1 + i_2}$,
$\phi(x^{i_1}x^{i_2}) = \phi(x^{i_1})\phi(x^{i_2}) \equiv y^{i_1}y^{i_2} = y^{i_1 + i_2}$なので、写像が定義できれば
準同型になりうる。


# 演習2.5.2

任意の$g,h \in G$を考える。$\phi_n(g)\phi_n(h) = g^n h^n$となる。一方で、Gは可換群なので、$\phi_n(gh) = (gh)^n = g^n h^n$となり、
$\phi_n(gh)=\phi_n(g)\phi_n(h)$となり、準同型になる。

# 演習2.5.3
## (1)
まず、$\phi(g) \in H$の位数がgの位数m以下であることを示す。

位数の定義2.4.1より、$g^m = 1_G$となる。命題2.5.3より、$\phi(g^m) = \phi(1_G) = 1_H$となる。

一方で、準同型写像なので、$\phi(g^m) = \phi(g)^m$となるので、$\phi(g)^m = 1_H$となり、
$\phi(g)$の位数はm以下になることがわかる。

$\phi(g)$の位数をnとすると、$m=nd+r,(d,r\in \mathbb{Z}, 0 \leq r < n)$となり、
$\phi(g)^m = \phi(g)^r = 1_H$となるが、位数の定義より、$r<n$であるrに対して、$\phi(g)^r = 1_H$となるのは
$r=0$にならないといけない。よって、nはmの約数となる。

## (2)
背理法で示す。

(1)より、$m=nd(d,m,n\in\mathbb{Z}, 0<n<m)$とし、$\phi$が単射のときに、mがnと等しくない、つまりd>1とする。

$g^m,g^n$を考えると、位数の定義より、$g^m=1_G$となるが、$g^n \neq 1_G$となり、$g^m \neq g^n$となる。

一方で、$\phi(g^m) = 1_H$で、かつ、$\phi(g^n) = \phi(g)^n = 1_H$となるので、$\phi(g^m) = \phi(g^n)$となるが、
$\phi$は単射とならず、矛盾する。

よって、$d = 1$となり、m=nとなる。

# 演習2.5.4
$\mathbb{Z}/n\mathbb{Z}$は定義2.2.1及び、例2.2.7を見ると、加法について、群となる。

$\mathbb{Z}/2\mathbb{Z}$の元の位数はたかだか、2なので(実際、\bar{0}の位数は1で、\bar{1}の位数は2)、
$\mathbb{Z}/2\mathbb{Z} \times \mathbb{Z}/2\mathbb{Z}$の元の位数もたかだか、2になる。
(実際、$(\bar{0}, \bar{0})$の位数は1で、$(\bar{0}, \bar{1})$が2, $(\bar{1}, \bar{0})$が2,$(\bar{1}, \bar{1})$が2になる。
元の要素の位数の最小公倍数が、その元の位数になる。)

一方で、$\mathbb{Z}/4\mathbb{Z}$について、演習2.4.7も参考にすると、$\bar{1},\bar{3}$の位数が4となる。($\bar{0}$は1,$\bar{2}$は2)

ここで、演習2.5.3(1)を振り返ると、$G=\mathbb{Z}/2\mathbb{Z} \times \mathbb{Z}/2\mathbb{Z}, H=\mathbb{Z}/4\mathbb{Z}$として、
$\phi_G : G \rightarrow H$が準同型であるとすると、$Im(\phi_G)$の位数は1か2でなければならない。つまり、$\bar{1},\bar{3}$は
$Im(\phi_G)$に含まれない。よって、$\phi_G$は全射でない。

演習1.1.5より、全単射であることと、逆写像を持つことが同値であることが示されている。

今回、$\phi_G$は全射でないことがわかったので、$\phi_G$は逆写像を持たない。よって、定義2.5.1(2)より、同型とは言えない。

ちなみに、ある、$g \in G$に対して、$\phi_G(g) = \bar{1} \in \mathbb{Z}/4\mathbb{Z}$だとする。

gの位数はたかだか、2で$\phi_G$は準同型のなので、$\phi_G(g)\phi_G(g) = \phi_G(g^2) = \phi_G(\bar{0}) = \bar{0}$。
一方で、$\phi_G(g)\phi_G(g) = \bar{1}\bar{1} = \bar{2}$となり、等しくならず、矛盾する。

n=0のとき、式2.1.3を参考に、$(xyx^{-1})^0 = 1_G$で$xy^0x^{-1} = x1_Gx^{-1} = xx^{-1} = 1_G$となり、
$(xyx^{-1})^0 =　xy^0x^{-1}$が言える。

# 演習2.5.6
$g \in G$が存在することを示す。以下のように仮定すると
$$
g=
\begin{pmatrix}
a & b\\
c & d\\
\end{pmatrix},
g^{-1}=
\frac{1}{det g}
\begin{pmatrix}
d & -b\\
-c & a\\
\end{pmatrix}
$$
定義2.5.18(2)より、$A = gBg^{-1}$となるとすると、
$$
\begin{pmatrix}
1 & 1\\
0 & 1\\
\end{pmatrix} = 
\frac{1}{ad-bc}
\begin{pmatrix}
a & b\\
c & d\\
\end{pmatrix}
\begin{pmatrix}
1 & 0\\
1 & 1\\
\end{pmatrix}
\begin{pmatrix}
d & -b\\
-c & a\\
\end{pmatrix}
$$
gは正則なので、
$$
\begin{pmatrix}
ad-bc & ad-bc\\
0     & ad-bc\\
\end{pmatrix} =
\begin{pmatrix}
ad-bc+bd & -b^2\\
d^2      & ad-bc-bd\\
\end{pmatrix} 
$$
これから、d=0,bd=0,ad-b(b-c)=0。つまり、d=0, b(b-c)=0。ただし、正則なので$ad - bc \neq 0$。
よって、$d = 0,b = c \neq 0$ 
## (1)
上記を考慮すると、aは任意、b=c,d=0となる、$a,b,c,d \in \mathbb{R}$が存在する。

例えば、a=0, b=c=1とすればよい。

よって、A,Bは$GL_2(\mathbb{R})$で共役になる。

## (2)
$g \in SL_2(\mathbb{R})$のときに、上記を考慮する。
すると、(1)に対して、$det g = ad - bc = 1$の条件が追加になる。

d=0,b=cなので、$b^2=-1$が必要になる。$b \in \mathbb{R}$であるので、これを満たすbはない。

よって、A,Bは$SL_2(\mathbb{R})$で共役にならない。

## (3)
$g \in SL_2(\mathbb{C})$のときに、上記を考慮する。

$b \in \mathbb{C}$であれば、$b^2=-1$を満たすは存在する。

例えば、b=c=iは条件を満たす。

よって、A,Bは$SL_2(\mathbb{C})$で共役になる。

# 演習2.5.7
Gを有限位数nの巡回群として$AutG \cong (\mathbb{Z}/n\mathbb{Z})^{\times}$となることを示す。

GからGへの恒等写像は自己同型になるので、$AutG \neq \emptyset$。

Gの生成元をgとして、Gの元hは$h=g^d(0 \leq d < n, d \in \mathbb{Z})$となる。($G = \{1(=g^0), g, \cdots, g^{n-1}\}$)

このとき、$\psi$を以下のように定義する。
$$
\psi: (\overline{k} \in (\mathbb{Z}/n\mathbb{Z})^{\times}) \rightarrow \phi_{\overline{k}} = h^{k} \in AutG (0 \leq k < n)
$$
このとき、演習2.5.2より、$\phi$は$G \rightarrow G$の準同型になるので、自己同型になりうる。

$\phi_{\overline{k}}(h) = h^k = (g^d)^k = g^{kd} = (g^k)^d$であり、系2.4.14より、kはnと互いに素になる。
すると、命題2.4.18より、$g^k$の位数はnとなり、命題2.4.19も参考にすると、$(g^k)^d(0 \leq d < n)$は互いに異なり、
$|G| = n = |\phi_{\overline{k}}(G)|$となるし、$\phi_{\overline{k}}(h)$は単射になる。
$\phi_{\overline{k}}(h) = g^{kd} \in G$なので、$\phi_{\overline{k}}(G) \subset G$で命題1.1.6より、
$\phi_{\overline{k}}(G) = G$になるし、$\phi_{\overline{k}}(h)$は全単射になる。つまり、$\phi_{\overline{k}} \in AutG$になる。
(well-definedになる。)

もし、$\overline{j} \in \mathbb{Z}/n\mathbb{Z}$(乗法群に限らない。)とすると、
生成元gに対して、$\phi_{\overline{j}}(g) = g^{j}$となり、G全体になる。もし、$\omega(g)$が自己同型(当然、準同型になる。)があり、
$\omega(g) = g^j = \phi_{\overline{j}}(g)$だとすると、命題2.5.12より、$\omega(g) = \phi_{\overline{j}}(g)$になる。
つまり、$AutG \subset \phi_{\overline{j}}(\overline{j} \in \mathbb{Z}/n\mathbb{Z})$

もし、$\overline{k} \in (\mathbb{Z}/n\mathbb{Z})^{\times}$でない、jで自己同型になるとする。
j=0のときは$\phi_{\overline{0}}(h)=1$となり、単射でなく、自己同型にならない。
また、jがnと互いに素でないとする。$n=je(0 < e \leq n, e \in \mathbb{Z})$なので、$\phi_{\overline{0}}(g^e)=1$であり、かつ、
$\phi_{\overline{j}}(g^e)=1$なので、これも単射でなく、自己同型にならない。
よって、$|AutG| = |(\mathbb{Z}/n\mathbb{Z})^{\times}|$

更に、先に記載したように、$\phi_{\overline{k}}(g) = g^{k}$はjが異なると、異なる。よって、$\psi$は単射になる。

これらから、命題1.1.6の(2)を用いて、$\psi$は全単射写像になる。

乗法群$(\mathbb{Z}/n\mathbb{Z})^{\times}$は当然、群の一部になる。$x, y \in (\mathbb{Z}/n\mathbb{Z})^{\times}$について、
$\phi_{\overline{xy}}(h) = h^{xy} = (h^y)^x = \phi_{\overline{x}}(\phi_{\overline{y}}(h)) = (\phi_{\overline{x}}\phi_{\overline{y}})(h)$となり、準同型になる。($\psi$が準同型になる。)

よって、$AutG \cong (\mathbb{Z}/n\mathbb{Z})^{\times}$が示された。

これを踏まえて、各問を考える。剰余群なので、生成元を1とする巡回群になっている。

以下では、特に剰余群について、考えてみる。

$\mathbb{Z}/n\mathbb{Z}$は命題2.2.9にあるように可換環になる。つまり、$\mathbb{Z}/n\mathbb{Z}$は加法について群になっており、
$\overline{0}$が単位元になる。

よって、定義2.2.1(3)にあるように、$a(b+c)=ab+ac$なので、$\phi_{\overline{k}}(x)=\overline{k}x$とすると、
$\phi_{\overline{k}}(x+y) = \phi_{\overline{k}}(x) + \phi_{\overline{k}}(y)$となり、加法の群で準同型になっていることがわかる。
つまり、$\phi_{\overline{k}}(x)$はwell-definedになっている。(もちろん、可換環なので、この演算はGの任意の元に対して、実施でき、
かつ、演算結果はGに含まれる。)

また、例2.3.18にあるように$\mathbb{Z}/n\mathbb{Z}$は$\overline{1}$を生成元とする、巡回群になる。
このとき、$0 \leq k < n$に対して、$\overline{k}\overline{1}=\overline{k}$となり、$\phi_{\overline{k}}$で、$\overline{1}$を、
任意の$\overline{j} \in G$に対応できる。

さて、$\phi_{\overline{k}}$でない、$\psi$があり、$\phi_{\overline{k}}(\overline{1}) = \psi(\overline{1}) = \overline{k}$とする。
$\mathbb{Z}/n\mathbb{Z}$は$\overline{1}$を生成元とする、巡回群なので、命題2.5.12より、$\phi_{\overline{k}}=\psi$となり、
準同型になるのは、$\phi_{\overline{k}}$に限られる。

ここまででは、$\overline{k} \in \mathbb{Z}/n\mathbb{Z}$になっている。しかし、実際は、$(\mathbb{Z}/n\mathbb{Z})^{\times}$となる。
まず、同型の定義2.5.1(2)より、準同型が、逆写像を持つ必要がある。P.4にあるように、逆写像を持つことと、全単射であることは同値である。
まず、$k=0$のとき、すべての元$\overline{a}$が$\overline{0}\overline{a} = \overline{0}$となり、単射にならないため、全単射にならない。
更に、$(\mathbb{Z}/n\mathbb{Z})^{\times}$でない元、$\overline{k}$を考えると、系2.4.14より、nとkが互いに素でなく、$n=kd(d \in Z, 0<d<n)$と
なる。そのため、$\overline{k}\overline{0} = \overline{0}$となり、かつ、$\overline{k}\overline{d} = \overline{0}$となり、
これも単射でない。
そのため、$AutG \subset \{ \phi_{\overline{k}}\}(k \in (\mathbb{Z}/n\mathbb{Z})^{\times})$となる。

一方で、$(\mathbb{Z}/n\mathbb{Z})^{\times}$は逆元を持つので、$\overline{k}$の逆元、$(\overline{k})^{-1}$が存在し、
$\phi_{\overline{k}^{-1}}$が逆写像になる。よって、$\phi_{\overline{k}}$は自己同型になる。

つまり、$AutG = \{ \phi_{\overline{k}}\}$。当然、要素数も等しい。

$\psi : k \rightarrow \phi_{\overline{k}} = h^k$となり、gに対して、考えると、値が異なるので、単射になる。

よって、$\psi$は同型になる。つまり、$AutG \cong (\mathbb{Z}/n\mathbb{Z})^{\times}$が示された。


# 演習2.5.8
## (1)
g=aとすると、baに対して、
$$
g(ba)g^{-1} = abaa^{-1} = ab
$$
となる。よって、共役になる。
(abに対しても、g'=bとすれば良い。)
## (2)
演習2.5.5を踏まえると
$$
(ab)^k = g (ba)^kg^{-1}
$$
baの位数がm、abの位相がnだとする。
$(ba)^m = 1_G$なので、$(ab)^m = g(ba)^mg^{-1} = g1_Gg^{-1} = gg^{-1} = 1_G$となるので、abの位数は$n \leq m$以下になる。

同様に
$$
(ba)^k = g' (ba)^kg'^{-1}
$$
$(ab)^n = 1_G$なので、$(ba)^n = g'(ab)^ng'^{-1} = g'1_Gg'^{-1} = g'g'^{-1} = 1_G$となるので、baの位数は$m \leq n$以下になる。

よって、$m=n$となり、ab,baの位数が等しくなる。

どちらかの位数が無限だとすると、もう一方が有限だと矛盾するので、ともに無限となり位数が等しくなる。

# 演習2.5.9
