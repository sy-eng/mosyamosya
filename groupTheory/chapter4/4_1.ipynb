# 4.1.7

まず、$a = ||x|| = ||y||$と置く。

命題4.1.20や系4.1.22を考え、$G = SO(n)$として、$x,y \in G(a, 0, ..., 0)^T, (a, 0, ..., 0)^T \in \mathbb{R}^n$となれば、
$(a, 0, ..., 0)$を代表元とする軌道ということができる。

$A \in G = SO(n), x^T = (x_1, x_2, \cdots, x_n)$として、
$$
A = 
\begin{pmatrix}
\frac{x_1}{a} & \cdots\\
\frac{x_2}{a} & \cdots\\
\vdots & \vdots\\
\frac{x_n}{a} & \cdots\\
\end{pmatrix}
$$
とすれば、$A(a, 0, ..., 0)^T=x$となる。

Aが$SO(n)$に含まれていることを確認する。
1列目はノルムは1になる。
2列目以降のi列は、その列より前の列に対して、直交するように設定できる。
n+1-i個の要素はすべての要素が0にならないようにすれば、自由に設定できる。
直交するようにしたあと、その列のノルムで割ることで、列のノルムを1にできる。

よって、Aが存在するのでxは(a, 0, ..., 0)のSO(n)による軌道に含まれる。同様にyも含まれる。

上記より、x,yはともに(a, 0, ..., 0)のSO(n)による軌道に含まれるので、SO(n)による同じ軌道に含まれる。

# 4.1.10
## (1)
四元数群において、1,-1は他のどの元とも可換になる。よって、それらの共役類はそれだけになる。

定義4.1.27の記号を考え、x = iに関して、共役類を考えると、g = 1, -1だと、y=iになる。
g = i, -iのときも、iの巡回群の要素なので可換となり、y=iとなる。
g = jのとき、$y = j i (-j) = - j k = -i$となる。
g = -jのとき、$y = -j i j = - j k = -i$となる。
g = kのとき、$y = k i (-k) = k j = -i$となる。
g = -kのとき、$y = -k i (k) = k j = -i$となる。
よって、iを含む共役類は$\{ i, -i\}$となる。

同様に考えると、すべての共役類は$\{ 1 \}, \{ -1 \}, \{ i, -i \}, \{ j, -j \}, \{ k, -k \}$となる。
(すべての元はどれかの共役類に1回だけ含まれる。)

このとき、共役類の完全代表系は共役類から1個ずつ取り出せば良いので、例えば、$\{\pm 1, i, j, k \}$となる。
