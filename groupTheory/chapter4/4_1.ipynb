# 4.1.1
Gは以下のようになっている。
$$G = \{ x_1 = 1, x_2 = (1 \: 2)(3 \: 4), x_3 = (1 \: 3)(2 \: 4), x_4 = (1 \: 4)(2 \: 3)\}$$
GのGへの作用を例4.1.14のように考える。

$g = x_2 = (1 \: 2)(3 \: 4)$とすると、$g x_1 = x_2, g x_2 = 1, 
g x_3 = (1 \: 2)(3 \: 4)(1 \: 3)(2 \: 4) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
=(1 \: 4)(2 \: 3) = x_4,
g x_4 = (1 \: 2)(3 \: 4)(1 \: 4)(2 \: 3) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
= (1 \: 3)(2 \: 4) = x_3$となり、
$x_1$が$x_2$に、$x_2$が$x_1$に、$x_3$が$x_4$に、$x_4$が$x_3$に、移動している。
よって、$\rho(x_2) = (1 \: 2)(3 \: 4) = x_2$となる。

同様に$g = x_3 = (1 \: 3)(2 \: 4)$とすると、$g x_1 = x_3, 
g x_2 = (1 \: 3)(2 \: 4)(1 \: 2)(3 \: 4) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
=(1 \: 4)(2 \: 3) = x_4,
g x_3 = 1,
g x_4 = (1 \: 3)(2 \: 4)(1 \: 4)(2 \: 3) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
= (1 \: 2)(3 \: 4) = x_2$となり、
$x_1$が$x_3$に、$x_2$が$x_4$に、$x_3$が$x_1$に、$x_4$が$x_2$に、移動している。
よって、$\rho(x_3) = (1 \: 3)(2 \: 4) = x_3$となる。

同様に$g = x_4 = (1 \: 4)(2 \: 3)$とすると、$g x_1 = x_4, 
g x_2 = (1 \: 4)(2 \: 3)(1 \: 2)(3 \: 4) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
=(1 \: 3)(2 \: 4) = x_3,
g x_3 = (1 \: 4)(2 \: 3)(1 \: 3)(2 \: 4) = 
\begin{pmatrix}
1 & 2 & 3 & 4\\
4 & 3 & 2 & 1
\end{pmatrix}
\begin{pmatrix}
1 & 2 & 3 & 4\\
3 & 4 & 1 & 2
\end{pmatrix}
=\begin{pmatrix}
1 & 2 & 3 & 4\\
2 & 1 & 4 & 3
\end{pmatrix}
= (1 \: 2)(3 \: 4) = x_2$、
g x_4 = 1となり、$x_1$が$x_4$に、$x_2$が$x_3$に、$x_3$が$x_2$に、$x_4$が$x_1$に、移動している。
よって、$\rho(x_4) = (1 \: 4)(2 \: 3) = x_4$となる。

# 4.1.2
例4.1.14(2)のように考えれば良い。$g = (2 \: 3)$としたとき、
$g x_1 = (2 \: 3) = x_4, 
g x_2 = 
\begin{pmatrix}
1 & 2 & 3\\
3 & 1 & 2
\end{pmatrix}
= (1 \: 3 \: 2) = x_6,
g x_3 = 
\begin{pmatrix}
1 & 2 & 3\\
2 & 3 & 1
\end{pmatrix}
= (1 \: 2 \: 3) = x_5,
g x_4 = 1_{\mathfrak{S}_3} = x_1,
g x_5 = (2 \: 3)(1 \: 2 \: 3)
= (1 \: 3) = x_3,
g x_6 = (2 \: 3)(1 \: 3 \: 2)
= (1 \: 2) = x_2
$

よって、$x_1$が$x_4$に、$x_2$が$x_6$に、$x_3$が$x_5$に、$x_4$が$x_1$に、$x_5$が$x_3$に、$x_6$が$x_2$に移る。
そのため、$\rho((2 \: 3)) = (1 \: 4)(2 \: 6)(3 \: 5)$となる。

# 4.1.4
例4.1.16に関して、$\sigma = (1 \: 2 \: 3) = x_5$としたときに、どうなるかを考える。

$Ad(\sigma)(x_1) = Ad(\sigma)(1) = 1, 
Ad(\sigma)(x_2) = Ad(\sigma)((1 \: 2)) = (1 \: 2 \: 3)(1 \: 2)(1 \: 3 \: 2) = (1 \: 2 \: 3)(1 \: 3) = (2 \: 3) = x_4,
Ad(\sigma)(x_3) = Ad(\sigma)((1 \: 3)) = (1 \: 2 \: 3)(1 \: 3)(1 \: 3 \: 2) = (1 \: 2 \: 3)(2 \: 3) = (1 \: 2) = x_2,
Ad(\sigma)(x_4) = Ad(\sigma)((2 \: 3)) = (1 \: 2 \: 3)(2 \: 3)(1 \: 3 \: 2) = (1 \: 2 \: 3)(1 \: 3) = (1 \: 3) = x_3,
Ad(\sigma)(x_5) = Ad(\sigma)((1 \: 2 \: 3)) = (1 \: 2 \: 3)(1 \: 2 \: 3)(1 \: 3 \: 2) = (1 \: 2 \: 3) = x_5,
Ad(\sigma)(x_6) = Ad(\sigma)((1 \: 3 \: 2)) = (1 \: 2 \: 3)(1 \: 3 \: 2)(1 \: 3 \: 2) = (1 \: 3 \: 2) = x_6
$
よって、$x_1,x_5,x_6$が変わらず、$x_2$が$x_4$に、$x_3$が$x_2$に、$x_4$が$x_3$になる。
したがって、$\rho((1 \: 2 \: 3)) = (2 \: 4 \: 3)$
