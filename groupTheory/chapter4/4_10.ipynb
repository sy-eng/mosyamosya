# 4.10.1
定理4.10.4(2)の証明を参考にする。P.141にあるように正八面体群と立方体群は同型であることに注意する。その時の対応についても踏まえて考える。

$F_1, F_2, F_3$を立方体の一つの頂点Pに集まる麺でこの順番に隣り合うように取る。(図はない。)$F'_1, F'_2, F'_3$をそれぞれ$F_1, F_2, F_3$に向かい合う面とし、
$C_1 = \{ F_1, F'_1 \}, C_2 = \{ F_2, F'_2 \}, C_3 = \{ F_3, F'_3 \}$と置く。Oは集合$\{ C_1, C_2, C_3 \}$に作用する。よって、置換表現$\rho : O \rightarrow \mathfrak{S}_3$が定まる。
すると、$F_1, F'_1$の中心を軸とする回転は$C_2 \rightarrow C_3 \rightarrow C_2 \rightarrow C_3 \rightarrow C_2$という置換を引き起こす。したがって、$(2 3) \in \rho(O)$である。

次に点P（とその対角の点を通る軸)周りの回転させるとこの元はOの元であり、$F_1 \rightarrow F_2 \rightarrow F_3 \rightarrow F_1$と置換する。
よって、この元は$C_1 \rightarrow C_2 \rightarrow C_3 \rightarrow C_1$という置換を引き起こす。
したがって、$(1 2 3) \in \rho(O)$である。

演習2.3.9(2)より$<(2 3), (1 2 3) = (2 3 1)> = \mathfrak{S}_3$なので全射になる。

なお、$|\mathfrak{S}_3| = 6$なので、具体的に全射であることを確認することも容易である。
上記のように$(1 2 3) \in \rho(O)$を確認したが、同じ回転で$(1 3 2) \in \rho(O)$を確認できる。
$(2 3) \in \rho(O)$を確認したが、$F_1$でなく$F_2, F_3$に対して同様の回転を確認すると$(1 2), (1 3) \in \rho(O)$を確認できる。
$1 \in \rho(O)$なので全射であることを確認できる。
