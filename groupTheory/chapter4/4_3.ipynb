## (1)
$n = 1$を考える。このとき、$N_1 = \{1\}$となり、これは命題2.3.2を考えると(1)1は単位元になり、(2)1のみの演算は1になり、$xy \in N_1$になり、(3)1のみの演算しかなく、1自体が逆元になり、$1 \in N_1$となる。よって、n=1のとき、$N_1$は$GL_n(\mathbb{C})$の部分集合となる。

n>1のとき、n=kまでで、$N_i$たちが部分群になっているとする。
$i = n - 1$のときは$N_{n-1} = \{I_n\}$となり、$n = 1$のときと同様、命題2.3.2を考えると$N_{n-1}$が部分群になることがわかる。

$i < n-1$を考える。
n=kでの$N_i$の元を$(N_i)_k$とすると
$$
A = {(N_i)_k}_A =
\begin{pmatrix}
1 & {n_i}_A\\
0 & {(N_i)_{k-1}}_A
\end{pmatrix},
B = {(N_i)_k}_B =
\begin{pmatrix}
1 & {n_i}_B\\
0 & {(N_i)_{k-1}}_B
\end{pmatrix}
$$
とする。このとき、
$$
AB =
\begin{pmatrix}
1 & {n_i}_AB\\
0 & {(N_i)_{k-1}}_AB
\end{pmatrix}
$$
となる。
右下の項は帰納法の仮定より、群となっているし、${(N_i)_{k-1}}_A$に対して、${(N_i)_{k-1}}_B$を決めることができる。
右上の項は上三角であることを考えると${n_i}_AB$となることがわかる。右上の項について、${(N_i)_{k-1}}_A, {(N_i)_{k-1}}_B, {n_i}_A$が与えられたときに、
${n_i}_B$は余因子行列を考えると${n_i}_B$のj番目の要素は${\delta_A}_{(j+1), 1}$となるが、$j \leq i$の間は余因子が0になる。なお、|A|=1。
よって、i-1までは0になり、${n_i}_B$が成り立つ。

上記より、命題2.3.2の(1)-(3)が成立することがわかるので、n=k+1でも$N_i$たちが部分群になることを確認できる。

## (2)
$p \in N_1, q \in N_i$を考えると
$$
p =
\begin{pmatrix}
1 & p_{1, 2} & \dots & \dots & \dots      & \dots & p_{1, n}  \\
  & \ddots   & \ddots&       &            &       & \vdots    \\
  &          & \ddots& \ddots&            &       & \vdots    \\
  &          &       & \ddots& \ddots     &       & \vdots    \\
  &          &       &       & \ddots     & \ddots& \vdots    \\
  & \huge{0} &       &       &            & \ddots& p_{n-1, n}\\
  &          &       &       &            &       & 1         
\end{pmatrix}
$$
$$
q =
\begin{pmatrix}
1 & 0        & \dots & 0     & q_{1, i+1} & \dots & q_{1, n}  \\
  & \ddots   & \ddots&       & \ddots     & \ddots& \vdots    \\
  &          & \ddots& \ddots&            & \ddots& q_{n-i, n}\\
  &          &       & \ddots& \ddots     &       & 0         \\
  &          &       &       & \ddots     & \ddots& \vdots    \\
  & \huge{0} &       &       &            & \ddots& 0         \\
  &          &       &       &            &       & 1         
\end{pmatrix}
$$
このとき、
$$
pq = 
\begin{pmatrix}
1 & p_{1, 2} & \dots & p_{1, i} & p_{1, i+1} + q_{1, i+1} & s_{1, i+2} & \dots  & s_{1, n}               \\
  & \ddots   & \ddots&          & \ddots                  & \ddots     & \ddots & \vdots                 \\
  &          & \ddots& \ddots   &                         & \ddots     & \ddots & s_{n-i-1,n}            \\
  &          &       & \ddots   & \ddots                  &            & \ddots & p_{n-i, n} + q_{n-i, n}\\
  &          &       &          & \ddots                  & \ddots     &        & p_{n-i-1, n}           \\
  &          &       &          &                         & \ddots     & \ddots & \vdots                 \\
  & \huge{0} &       &          &                         &            & \ddots & p_{n-1, n}             \\
  &          &       &          &                         &            &        & 1         
\end{pmatrix}
$$
$$
qp = 
\begin{pmatrix}
1 & p_{1, 2} & \dots & p_{1, i} & p_{1, i+1} + q_{1, i+1} & t_{1, i+2} & \dots  & t_{1, n}               \\
  & \ddots   & \ddots&          & \ddots                  & \ddots     & \ddots & \vdots                 \\
  &          & \ddots& \ddots   &                         & \ddots     & \ddots & t_{n-i-1,n}            \\
  &          &       & \ddots   & \ddots                  &            & \ddots & p_{n-i, n} + q_{n-i, n}\\
  &          &       &          & \ddots                  & \ddots     &        & p_{n-i-1, n}           \\
  &          &       &          &                         & \ddots     & \ddots & \vdots                 \\
  & \huge{0} &       &          &                         &            & \ddots & p_{n-1, n}             \\
  &          &       &          &                         &            &        & 1         
\end{pmatrix}
$$
ただし、$s_{a, b} = p_{a, b} + q_{a, b} + \sum_{c = a + 1}^{b - i} p_{a, c} q_{c, b}$
(一般に$s_{a, b} = \sum_{k=1}^{n} p_{a, k} q_{k, b}$。上三角行列なので、$a > b$のとき、$p_{a, b}, q_{a, b} = 0$。
また、対角が1なので、$p_{j, j}, q_{j, j} = 1$,$a < b$で$b < a + i, q_{a, b} = 0$)
同様に$t_{a, b} = p_{a, b} + q_{a, b} + \sum_{c = a + i}^{b - 1} p_{c, b} q_{a, c}$

さて、$[N_1, N_i] \subset N_{i+1}$を示したいわけだが、これを示すために$N_i \triangleleft N_1$を示す。これを示すことで、命題4.3.4の証明で議論されたように、
$[N_1, N_i] \subset N_{i+1}$は$\forall x \in N_1, \forall y \in N_i, xy N_{i+1} = yx N_{i+1}$と同値になる。

$N_i \triangleleft N_1$であることは定義2.8.1より$\forall p \in N_1, \forall q \in N_i, \exists q' \in N_i, pqp^{-1} = q'$つまり、$pq=q'p$となるq'が存在することを示せば良い。
q'を含む計算に出てくる、q'に依存する変数には'をつけるとするし、上で計算した行列計算の結果を踏まえるとまず、$p_{1, i+1} + q_{1, i+1}$から$p_{n-i, n} + q_{n-i, n}$の
斜めの要素を考えると$pq=q'p$となるには、$q_{1, i+1} = q'_{1, i+1}$などとすればよい。
ここから、今求めたところの一つ右側の要素を考えると、$s_{1, i+2} = t'_{1, i+2}$などとなる。sはp,qが与えられると値が決まり、t'の定義を踏まえると、
$q'_{a, b} = s_{a, b} - p_{a, b} - \sum_{c = a + i}^{b - 1} p_{c, b} q'_{a, c}$となる。b-1列までの$q'_{a, c}$はわかっているので、$q'_{a, b}$が求まる。
このように左から順に求めていくとq'の各要素を決められるので、q'は存在する。よって、$N_i \triangleleft N_1$。

$[N_1, N_i] \subset N_{i+1}$を示すために、$\forall x \in N_1, \forall y \in N_i, xy N_{i+1} = yx N_{i+1}$を示す。
つまり、$\forall p \in N_1, \forall q \in N_i, \exists h \in N_{i+1}, pq = qph$(これは命題4.3.4付近で議論した。)
hが必ず存在することを、正規部分群であることを示した方法と同様に示す。
pq,qrの計算で、s,tでない変数を見ると、同じであることがわかる。つまり、左下からp+qとなっているところまでは同じであることがわかる。
qrhを計算すると
$$
qph =  
\begin{pmatrix}
1 & p_{1, 2} & \dots & p_{1, i} & p_{1, i+1} + q_{1, i+1} & u_{1, i+2} & \dots  & u_{1, n}               \\
  & \ddots   & \ddots&          & \ddots                  & \ddots     & \ddots & \vdots                 \\
  &          & \ddots& \ddots   &                         & \ddots     & \ddots & u_{n-i-1,n}            \\
  &          &       & \ddots   & \ddots                  &            & \ddots & p_{n-i, n} + q_{n-i, n}\\
  &          &       &          & \ddots                  & \ddots     &        & p_{n-i-1, n}           \\
  &          &       &          &                         & \ddots     & \ddots & \vdots                 \\
  & \huge{0} &       &          &                         &            & \ddots & p_{n-1, n}             \\
  &          &       &          &                         &            &        & 1         
\end{pmatrix}
$$
hの(a, b)成分をh(a,b)、qpの(a, b)成分をr(a,b)とも表すと
$u_{a, b} = r_{a, b} + h_{a, b} + \sum_{c = a + 1}^{b - i} r_{a, c} h_{c, b} = t_{a, b} + h_{a, b} + \sum_{c = a + 1}^{b - i} r_{a, c} h_{c, b}$
特にp+qとなっているところの一つ上の要素は$u_{a, b} = t_{a, b} + h_{a, b}$となる。
p+qの一つ上の要素から検討すると、$h_{a,b} = t_{a,b} - s_{a,b}$とすれば良く、hを決定できる。同様にそれぞれのひとつ上を考えると、
$h_{a, b} = s_{a, b} - t_{a, b} - \sum_{c = a + 1}^{b - i} r_{a, c} h_{c, b}$となり、a+1行目以下はわかっているので、hの要素を決定できる。
よって、hは存在する。

これにより、$[N_1, N_i] \subset N_{i+1}$が示された。

## (3)
(2)で$N_i \triangleleft N_1$を示した。
また、$[N_1, N_i] \subset N_{i+1}$を示し、命題4.3.4の議論を踏まえると$G_i/G_{i+1}$は$G/G_{i+1}$の中心に含まれることがわかる。
定義4.3.3(2)より、べき零群であると言える。

命題4.3.10の証明について考える。

命題4.3.9の証明と同様に考える。

同値であることを言うために群Gがべき零群であるときに$Z_n (G) = \{ 1 \}$となることを示し、その後、$Z_n (G) = \{ 1 \}$となれば、群Gがべき零群であることを示す。

まず、群Gがべき零群であるときに$Z_n (G) = \{ 1 \}$となることを示す。

$G = G_0 \supset G_1 \supset \cdots \supset G_n = \{ 1 \}$を定義4.3.3(2)の部分群の列とする。$G_i / G_{i+1}$は$G / G_{i+1}$の中心に含まれるので、
命題4.3.4(1)の証明にもあるように$[G, G_i] \subset G_{i+1}$

命題4.3.10の直前にある定義より$Z_0(G) = G = G_0$なので、$i = 0$のとき、$Z_i(G) \subset G_i$が成り立っていることがわかる。

$i = k-1$までで$Z_i(G) \subset G_i$が成り立っているとする。

命題4.3.10の直前にある定義より$Z_k(G) = [G, Z_{k-1}(G)]$だが、帰納法の仮定より$Z_{k-1}(G) \subset G_{k-1}$なので、命題2.3.14も考慮して
$[G, Z_{k-1}(G)] \subset [G, G_{k-1}]$

ここまでの内容をまとめると、
$Z_i(G) = [G, Z_{i-1}(G)] \subset [G, G_{i-1}] \subset G_{i}$
よって、$Z_i(G) \subset G_i$

$G_n = \{ 1 \}$であり、$Z_k(G)$が空集合でないとすると、$Z_{k+1}(G) = [G, Z_{k}(G)]$は空集合にならないため、$Z_{k+1}(G)$は空集合にならない。
よって、$Z_i(G) = \{ 1 \}$となる。

逆に、$Z_n (G) = \{ 1 \}$となれば、群Gがべき零群であることを示す。

$G_0 = G = Z_0(G)$とし、$i = 1, \cdots, n$に対し、$G_i = Z_i(G)$とおく。$Z_0(G) = G, Z_1(G) = [G, G]$で命題4.3.2(1)より$Z_1(G) \subset Z_0(G)$。
その時に、$Z_{i+1}(G) = [G, Z_i(G)]$なので、$Z_{i+1}(G) \subset Z_i(G)$。すると、$Z_0(G) \supset Z_1(G) \supset \cdots \supset Z_n(G) = \{ 1 \}$,
$G = G_0 \supset G_1 \supset \cdots \supset G_n = \{ 1 \}$である。
また、$Z_{i+1}(G) = [G, Z_i(G)] \subset Z_i(G)$となり、任意の$g \in G, z \in Z_i(G)$に対して、$gzg^{-1}z^{-1} \in Z_i(G), gzg^{-1} \in Z_i(G)$となり、
定義2.8.1より、任意のiについて、$Z_i(G) \triangleleft G$となる。

部分群の列の定義を踏まえると$G_{i+1} = [G, G_i]$。命題4.3.4(1)の証明の冒頭のように$G_{i+1} \triangleleft G_i$。
($G_{i+1}$が$G_i$の部分群であること,$G_{i}$が$G$の部分群であることは定義から言える。また、Gの任意の元gが$gG_{i+1}g^{-1} \in G_{i+1}$となるので
定義2.8.1より、$G_{i+1} \triangleleft G$といえる。さらに、$G_i$の元gについても同様のことが言えるので、$G_{i+1} \triangleleft G_i$)

$G_{i+1} = [G, G_i]$からはさらに$G_i / G_{i+1}$の中心が$G / G_{i+1}$の中心であることも言える。
任意の$g \in G, g_i \in G_i$について、定義4.3.1(2)より、$g g_i g^{-1} {g_i}^{-1}$が$[G, G_i]$の生成元になることが言え、これらは$g g_i g^{-1} {g_i}^{-1} \in [G, G_i] = G_{i+1}$となる。
$G_{i+1}$は群なので、演算が閉じており、$g g_i g^{-1} {g_i}^{-1} G_{i+1} = G_{i+1}$。(群に元を左から作用させると考えると統合になる。)
さらに正規部分群であることも考慮すると、以下のように変形できる。
$g g_i G_{i+1} g^{-1} {g_i}^{-1} = G_{i+1}$, $g g_i G_{i+1} = G_{i+1} {g_i} g$, $g G_{i+1} g_i G_{i+1} = {g_i} G_{i+1} g G_{i+1}$
よって、p.58のように$G/G_{i+1},G_i/G_{i+1}$の元がそれぞれ、$g G_{i+1}, g_i G_{i+1}$となるので、$G_i / G_{i+1}$の中心が$G / G_{i+1}$の中心であるといえる。
