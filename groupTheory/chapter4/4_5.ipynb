Stab(y)に含まれるには、$\sigma 1 = 1, \sigma (1\,2)(3\,4) = (1\,2)(3\,4)$もしくは、$\sigma 1 = (1\,2)(3\,4), \sigma (1\,2)(3\,4) = 1$となる必要がある。

## (1)
p,qは<u>異なる</u>素数なので、p>qとしておく。逆だとしても、話は同様になる。

H,KをそれぞれGのシローp部分群、シローq部分群とする。|H|=p, |K|=qは素数なので、命題2.6.22を踏まえると巡回群となり、$H \cong \mathbb{Z}/p\mathbb{Z}, K \cong \mathbb{Z}/q\mathbb{Z}$である。
シローp部分群の数をs、シローq部分群の数をtとすると、定理4.5.7(4)より$s \equiv 1\,mod\,p, t \equiv 1\,mod\,q$である。
定理4.5.7(4)の証明を参考にすると$H \subset N_G (H), K \subset N_G (K)$なので、$[G:N_G(H)],[G:N_G(K)]$は定理2.6.20,2.6.21を踏まえると
$[G:N_G(H)]=|G|/|N_G(H)|, |N_G(H)| = m|H|, m \in \mathbb{N}, [G:N_G(H)]=|G|/(m|H|) = [G:H]/m$などとなり、それぞれ$[G:H] = q, [G:K] = p$の約数となる。
これらは素数なので、s,tの可能性はそれぞれ、1,qおよび1,pである。
p>qとしたので、<u>少なくとも</u>$q \not \equiv 1 \, mod \, p$なので、s=1である。定理4.5.6より$s = |G|/|N_G(H)| = 1$なので、$|G| = |N_G(H)|$,命題1.1.6(1)より$G = N_G(H)$。
よって、Hの共役はHのみである。また、定義2.8.1、4.1.26(1)より、$H \triangleleft G$を意味する。

Hは$\{ 1 \},G$以外の自明でない正規部分群になるが、これは定義4.3.11の単純群の定義から外れるので、Gは単純群でない。

## (2)
問題の条件が加わると、例4.5.8に沿って、(1)の話を進めることができる。

$p \not \equiv 1 \, mod \, q$が追加になるので、t=1である。よって、同様にKの共役はKのみである。これは、$K \triangleleft G$を意味する。

$|H \cap K|$は|H|=p,|K|=qの約数なので1である。よって$H \cap K = \{ 1_G \}$(演習2.6.4も参考のこと。)である。
$H,K \triangleleft G$なので、$HK \subset G$は部分群である。(命題2.10.3(1))H,Kは部分群で、$1_G$を含むので$HK \supset H, K$。命題2.6.21を考慮すると|HK|はp,qの公倍数である。
HKはGの部分群で、$|HK| \leq |G| = pq$なので、命題1.1.6(1)より、HK=Gとなる。命題2.9.2より$G \cong H \times K \cong \mathbb{Z}/p \mathbb{Z} \times \mathbb{Z}/q \mathbb{Z}$である。
定理2.9.3より、$G \cong \mathbb{Z}/pq \mathbb{Z}$である。

つまり、巡回群となる。
