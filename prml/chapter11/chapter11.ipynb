演習11.1を検討する。  
(11.1)の
$$
E[f] = \int f(\bf{z})p(\bf{z})d\bf{z}
$$
及び、(11.2)の
$$
\hat{f} = \frac{1}{L}\sum_{l = 1}^{L} f({\bf{z}}^{(l)})
$$
から$E[\hat{f}]$を検討する。${\bf{z}}^{(l)}$は、$p(z)$にしたがう、独立同分布になるので、
$$
E[\hat{f}] = \int \hat{f}\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \int \frac{1}{L}\sum_{l = 1}^{L} (f({\bf{z}}^{(l)}))\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \frac{1}{L}\sum_{l = 1}^{L} (\int f({\bf{z}}^{(l)})\prod_{k = 1}^{L} p({\bf{z}}^{(k)})d{\bf{z}}^{(k)})
$$
定数に対して、$\int p({\bf{z}}^{(k)})d{\bf{z}}^{(k)} = 1$を考慮すると、$l = k$となる部分以外が、1をかけることになり、$\int f({\bf{z}}^{(k)})p({\bf{z}}^{(k)})d{\bf{z}}^{(k)} = E[f]$であるため、
$$
E[\hat{f}] = \frac{1}{L}\sum_{l = 1}^{L} \int f({\bf{z}}^{(l)})p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \frac{1}{L}E[f]L = E[f]
$$
分散について検討する。(1.38),(1.39)より、
$$
var[\hat{f}] = E[{({\hat{f}} - E[\hat{f}])}^2] = E[{\hat{f}}^2] - {E[\hat{f}]}^2
$$
上記と同様に$E[{\hat{f}}^2]$を考慮する。
$$
E[{\hat{f}}^2] = \int {(\frac{1}{L}\sum_{l = 1}^{L} (f({\bf{z}}^{(l)})))}^2\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)}
= \frac{1}{L^2} \sum_{k}^{L} (\int {f({\bf{z}}^{(k)})}^2 d{\bf{z}}^{(k)}+ \int\int \sum_{l = 1, l \neq k}^{L} (f({\bf{z}}^{(l)})) f({\bf{z}}^{(k)}) d{\bf{z}}^{(l)}d{\bf{z}}^{(k)}) = \frac{1}{L^2}(E[f^2] + (L - 1){E[f]}^2)L = \frac{1}{L}(E[f^2] + (L - 1){E[f]}^2)
$$
よって、
$$
var[\hat{f}] = E[{\hat{f}}^2] - {E[\hat{f}]}^2 = \frac{1}{L}(E[f^2] + (L - 1){E[f]}^2) - {E[f]}^2 = \frac{1}{L}(E[f^2] - {E[f]}^2)=
\frac{1}{L}(E[f^2 - {E[f]}^2])= \frac{1}{L}(E[f^2 - 2fE[f]+ {E[f]}^2]) = \frac{1}{L}(E[{(f - E[f])}^2])
$$

演習11.2を検討する。

p(y) = p(z)
\begin{vmatrix}
\frac{dz}{dy}

\int_{-\infty}^{y} p(\hat{y})d\hat{y} = \int_{-\infty}^{z} p(\hat{z})d\hat{z} = \int_{0}^{z} d\hat{z} = z

よって、
$$
z = h(y) = \int_{-\infty}^{y} p(\hat{y})d\hat{y}

$$
よって、
$$
y = -\frac{1}{\lambda}ln(1 - z)
$$

$$
$$
z = z_i + \frac{p(z_i)}{{p(z_i)}'}ln(\frac{{p(z_i)}'}{{p(z_i)}^2}(y - \frac{\sum_j^{i-1} A_j}{A}) + {exp}^{\frac{{p(z_i)}'}{p(z_i)}(z_{il} - z_i)})
$$

```python
#Implementation of adaptive rejection sampling
#Here, Gam(z|a,b) = b^a*z^(a-1)exp(-bz) / gamma(a) is sampled.
#but, here, gamma(a) is unknown.
#\tilde{p} = z^(a-1)exp(-bz))
#ln (\tilde{p}) = -bz + (a -1)*ln z
#(ln (\tilde{p}))' = -b + (a-1)/z
#if a > 1, (ln (\tilde{p}))'' = -(a-1)/(z^2) < 0
#This shows ln (\tilde{p}) is a concave function.
#In this code a = 2 and b = 1. 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#import pandas as pd
from scipy.stats import gamma
import math

def pTilde(z):
    return z**(a-1)*np.exp(-b*z)

def differencePTilde(z):
    return (a-1)*(z**(a-2))*np.exp(-b*z) - b*(z**(a-1))*np.exp(-b*z)

def getBorderZ(z1, z2):
    a = min(z1, z2)
    b = max(z1, z2)
    aDiv = differencePTilde(a)/pTilde(a)
    bDiv = differencePTilde(b)/pTilde(b)
    pDiv = pTilde(b) / pTilde(a)
    return (aDiv*a - bDiv*b + math.log(pDiv)) / (aDiv - bDiv)

class Tangent:
    def __init__(self, z, z_min, z_max):
        self.z = z
        self.z_min = z_min
        self.z_max = z_max
        self.slope = differencePTilde(z)/pTilde(z)
        self.area = self.area()
        
    def line(self, z):
        return self.slope*(z - self.z) + math.log(pTilde(self.z))
    
    def area(self):
        return pTilde(self.z) / self.slope * (np.exp(self.slope*(self.z_max - self.z)) - np.exp(self.slope*(self.z_min - self.z)))
        
a = 20.0
b = 2.0

#In this code, z = (0, 5)
zSamplingArea = 30.0

figNum = 5
fig = plt.figure(figsize = (12.0, 12.0*figNum))
ax1 = fig.add_subplot(figNum, 1, 1, title = "z - pTilde(Initial)", xlim = (-0.1, zSamplingArea + 0.1))#, ylim = (0.0, 0.5))
ax2 = fig.add_subplot(figNum, 1, 2, title = "z - ln(pTilde)(Initial)", xlim = (-0.1, zSamplingArea + 0.1))#, ylim = (-5.0, 1.0))
ax3 = fig.add_subplot(figNum, 1, 3, title = "z - pTilde(Final)", xlim = (-0.1, zSamplingArea + 0.1))#, ylim = (0.0, 0.5))
ax4 = fig.add_subplot(figNum, 1, 4, title = "z - ln(pTilde)(Final)", xlim = (-0.1, zSamplingArea + 0.1))#, ylim = (-5.0, 1.0))
ax5 = fig.add_subplot(figNum, 1, 5, title = "z - gamma", xlim = (-0.1, zSamplingArea + 0.1))#, ylim = (0.0, 0.5))

initPoints = zSamplingArea * np.random.rand(2)
borderZ = getBorderZ(initPoints[0], initPoints[1])
tangents = [Tangent(min(initPoints), 0, borderZ), Tangent(max(initPoints), borderZ, zSamplingArea)]

# plot \tilde{p}
zBinNum = 100.0
zs = [z/zBinNum for z in range(1, int(zBinNum*zSamplingArea))]
ps = [pTilde(z) for z in zs]
ax1.plot(zs, ps)

#plot ln (\tilde{p})
lnps = [math.log(p) for p in ps]
ax2.plot(zs, lnps)
area = 0.0
tangentZ = [tangents[0].z_min]
tangentP = [tangents[0].line(tangents[0].z_min)]
for tangent in tangents:
    tangentZ.append(tangent.z_max)
    tangentP.append(tangent.line(tangent.z_max))
    area += tangent.area
ax2.plot(tangentZ, tangentP)
print(area)

#plot exp
index = 0
qs = []

for z in zs:
    while(z > tangents[index].z_max):
        index += 1
    qs.append(pTilde(tangents[index].z) * np.exp(tangents[index].slope*(z - tangents[index].z)))
ax1.plot(zs, qs)

samplingLen = 100000.0
sampledX = np.random.rand(int(samplingLen))
acceptedZ = []
print("max {}".format(max(sampledX)))

#print(tangents[0].area)
#print(tangents[1].area)
#print(area)

for x in sampledX:
    index = 0
    areaTmp = 0.0
    while((areaTmp + tangents[index].area) < area*x):
#        print("{} {} {}".format(index, areaTmp, x))
        areaTmp += tangents[index].area
        index += 1
#    print("area {} {} {}".format(areaTmp + tangents[index].area, area*x, areaTmp))
    zi = tangents[index].z
    zslope = tangents[index].slope
#    print("data {} {} {}".format(zslope/pTilde(zi)*(area*x - areaTmp) + np.exp(zslope * (tangents[index].z_min - zi)),
#                                 zslope/pTilde(zi)*(area*x - areaTmp), np.exp(zslope * (tangents[index].z_min - zi))))
    z = zi + math.log(zslope/pTilde(zi)*(area*x - areaTmp) + np.exp(zslope * (tangents[index].z_min - zi))) / zslope
    y = np.random.rand() * pTilde(zi) * np.exp(zslope*(z - zi))
    
    if y <= pTilde(z):
        acceptedZ.append(z)
    else:
#        print("rejected : {}".format(z))
        zTmp = tangents[index].z
        zMinTmp = tangents[index].z_min
        zMaxTmp = tangents[index].z_max
        borderZ = getBorderZ(zTmp, z)
        area -= tangents[index].area
        tangents.pop(index)
        if z < zTmp:
            if index == 0:
                tangents.insert(index, Tangent(z, zMinTmp, borderZ))
                tangents.insert(index + 1, Tangent(zTmp, borderZ, zMaxTmp))
            else:
                zTmp2 = tangents[index - 1].z
                zMinTmp2 = tangents[index - 1].z_min
                zMaxTmp2 = tangents[index - 1].z_max
                borderZ2 = getBorderZ(zTmp2, z)
                area -= tangents[index - 1].area
                tangents.pop(index - 1)
                tangents.insert(index - 1, Tangent(zTmp2, zMinTmp2, borderZ2))
                tangents.insert(index, Tangent(z, borderZ2, borderZ))
                tangents.insert(index + 1, Tangent(zTmp, borderZ, zMaxTmp))
                area += tangents[index - 1].area
            area += tangents[index].area
            area += tangents[index + 1].area
                
        if zTmp < z:
            if index == len(tangents):
                tangents.insert(index, Tangent(zTmp, zMinTmp, borderZ))
                tangents.insert(index + 1, Tangent(z, borderZ, zMaxTmp))
            else:
#                print("len {} {}".format(len(tangents), index))
                zTmp2 = tangents[index].z
                zMinTmp2 = tangents[index].z_min
                zMaxTmp2 = tangents[index].z_max
                borderZ2 = getBorderZ(z, zTmp2)
                area -= tangents[index].area
                tangents.pop(index)
                tangents.insert(index, Tangent(zTmp, zMinTmp, borderZ))
                tangents.insert(index + 1, Tangent(z, borderZ, borderZ2))
                tangents.insert(index + 2, Tangent(zTmp2, borderZ2, zMaxTmp2))
                area += tangents[index + 2].area
            area += tangents[index].area
            area += tangents[index + 1].area

##reject rate
print(1 - len(acceptedZ)/samplingLen)
print("maxZ {}".format(max(acceptedZ)))
#plt.hist(acceptedZ, bins = 300, range = (0, 30.0), density = 1.0)
#print(acceptedZ)

ax3.plot(zs, ps)
#plot exp
index = 0
qs = []

for z in zs:
    while(z > tangents[index].z_max):
        index += 1
    qs.append(pTilde(tangents[index].z) * np.exp(tangents[index].slope*(z - tangents[index].z)))
ax3.plot(zs, qs)

ax4.plot(zs, lnps)
tangentZ = [tangents[0].z_min]
tangentP = [tangents[0].line(tangents[0].z_min)]
for tangent in tangents:
    tangentZ.append(tangent.z_max)
    tangentP.append(tangent.line(tangent.z_max))
ax4.plot(tangentZ, tangentP)

pGamma = gamma.pdf(zs, a, scale = 1 / b)
ax5.plot(zs, pGamma)
#This histgram density is a little larger than gamma distribution.
#Because the histgram is normalized in [0, zSamplingArea]
ax5.hist(acceptedZ, bins = 100, range = (0, zSamplingArea), density = 1.0)

plt.show()
```

上記のように適応的棄却サンプリングでは棄却率が非常に小さくなる。  
最初は棄却率がそれほど小さくないが、途中から、棄却率が小さくなる。  
具体的には単純な棄却サンプリングが36%程度の棄却率に対して、適応的棄却サンプリングは0.01%以下と非常に小さい。
