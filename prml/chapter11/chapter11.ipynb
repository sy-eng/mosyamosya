演習11.1を検討する。  
(11.1)の
$$
E[f] = \int f(\bf{z})p(\bf{z})d\bf{z}
$$
及び、(11.2)の
$$
\hat{f} = \frac{1}{L}\sum_{l = 1}^{L} f({\bf{z}}^{(l)})
$$
から$E[\hat{f}]$を検討する。${\bf{z}}^{(l)}$は、$p(z)$にしたがう、独立同分布になるので、
$$
E[\hat{f}] = \int \hat{f}\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \int \frac{1}{L}\sum_{l = 1}^{L} (f({\bf{z}}^{(l)}))\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \frac{1}{L}\sum_{l = 1}^{L} (\int f({\bf{z}}^{(l)})\prod_{k = 1}^{L} p({\bf{z}}^{(k)})d{\bf{z}}^{(k)})
$$
定数に対して、$\int p({\bf{z}}^{(k)})d{\bf{z}}^{(k)} = 1$を考慮すると、$l = k$となる部分以外が、1をかけることになり、$\int f({\bf{z}}^{(k)})p({\bf{z}}^{(k)})d{\bf{z}}^{(k)} = E[f]$であるため、
$$
E[\hat{f}] = \frac{1}{L}\sum_{l = 1}^{L} \int f({\bf{z}}^{(l)})p({\bf{z}}^{(l)})d{\bf{z}}^{(l)} = \frac{1}{L}E[f]L = E[f]
$$
分散について検討する。(1.38),(1.39)より、
$$
var[\hat{f}] = E[{({\hat{f}} - E[\hat{f}])}^2] = E[{\hat{f}}^2] - {E[\hat{f}]}^2
$$
上記と同様に$E[{\hat{f}}^2]$を考慮する。
$$
E[{\hat{f}}^2] = \int {(\frac{1}{L}\sum_{l = 1}^{L} (f({\bf{z}}^{(l)})))}^2\prod_{l = 1}^{L} p({\bf{z}}^{(l)})d{\bf{z}}^{(l)}
= \frac{1}{L^2} \sum_{k}^{L} (\int {f({\bf{z}}^{(k)})}^2 d{\bf{z}}^{(k)}+ \int\int \sum_{l = 1, l \neq k}^{L} (f({\bf{z}}^{(l)})) f({\bf{z}}^{(k)}) d{\bf{z}}^{(l)}d{\bf{z}}^{(k)}) = \frac{1}{L^2}(E[f^2] + (L - 1){E[f]}^2)L = \frac{1}{L}(E[f^2] + (L - 1){E[f]}^2)
$$
よって、
$$
var[\hat{f}] = E[{\hat{f}}^2] - {E[\hat{f}]}^2 = \frac{1}{L}(E[f^2] + (L - 1){E[f]}^2) - {E[f]}^2 = \frac{1}{L}(E[f^2] - {E[f]}^2)=
\frac{1}{L}(E[f^2 - {E[f]}^2])= \frac{1}{L}(E[f^2 - 2fE[f]+ {E[f]}^2]) = \frac{1}{L}(E[{(f - E[f])}^2])
$$

演習11.2を検討する。  
(11.5)の
$$
\begin{equation}
p(y) = p(z)
\begin{vmatrix}
\frac{dz}{dy}
\end{vmatrix}
\end{equation}
$$
を考慮し、zが(0,1)で一様分布なので、
$$
\int_{-\infty}^{y} p(\hat{y})d\hat{y} = \int_{-\infty}^{z} p(\hat{z})d\hat{z} = \int_{0}^{z} d\hat{z} = z
$$
よって、
$$
z = h(y) = \int_{-\infty}^{y} p(\hat{y})d\hat{y}

P.240の指数分布について、詳細に検討する。
$$
p(y) = \lambda exp(-\lambda y)
$$
なので、
$$
h(y) = \int_{0}^{y} p(\hat{y}) d\hat{y} = \int_{0}^{y} \lambda exp(-\lambda \hat{y}) d\hat{y} = {[-exp(-\lambda y)]}_{0}^{y} = 1 - exp(-\lambda y) 
$$
よって、
$$
y = -\frac{1}{\lambda}ln(1 - z)
$$

z = \int_{-\frac{\pi}{2}}^{\theta} \frac{1}{\pi}\frac{1}{1+{tan}^2\hat{\theta}} \frac{1}{{cos}^2 \hat{\theta}} d\hat{\theta} = \int_{-\frac{\pi}{2}}^{\theta} \frac{1}{\pi}d\hat{\theta} = \frac{1}{\pi}[\theta]_{-\frac{\pi}{2}}^{\theta} = \frac{\theta}{\pi} + \frac{1}{2}

$$
y_2 = z_2 {(\frac{-2\, ln\, r^2}{r^2})}^{1/2} = r sin \theta {(\frac{-2\, ln\, r^2}{r^2})}^{1/2} = sin \theta {(-2\, ln\, r^2)}^{1/2} = 2 sin \theta {(- ln\, r)}^{1/2}
$$
ここで、
$$
\begin{equation}
\begin{vmatrix}
\frac{\partial (z_1, z_2)}{\partial (r, \theta)}
\end{vmatrix}
=

cos \theta & -r sin \theta \\

= r
$$
$$
\begin{equation}
\begin{vmatrix}
\frac{\partial (y_1, y_2)}{\partial (r, \theta)}
\end{vmatrix}
=
\begin{vmatrix}
-\frac{cos \theta}{r\sqrt{-ln\, r}} & -2sin \theta \sqrt{-ln\, r}\\

\end{vmatrix}
=
\begin{vmatrix}
-\frac{2}{r}

$$
改めて、考えると、
$$
p(y_1, y_2)
\begin{equation}
\begin{vmatrix}
\frac{\partial (y_1, y_2)}{\partial (r, \theta)}
\end{vmatrix}

p(acceptance | z) = \int_0^{\tilde{p}(z)} \frac{1}{kq(z)} dz = \frac{\tilde{p}(z)}{kq(z)}

p(acceptance, z) = p(acceptance | z)q(z) = \frac{\tilde{p}(z)}{kq(z)}q(z) = \frac{\tilde{p}(z)}{k}
$$

$$
p(acceptance) = \int p(acceptance, z) dz = \frac{1}{k}\int \tilde{p}(z) dz

$$
p(z) = p(z | acceptance) = \frac{p(acceptance, z)}{p(acceptance)} = \frac{\frac{\tilde{p}(z)}{k}}{\frac{1}{k}\int \tilde{p}(z) dz} = \frac{\tilde{p}(z)}{\int \tilde{p}(z) dz}

kは任意の定数なので、kb→kとし直せば、成立する。
