P258のすべての要素が正であれば、ある状態遷移を繰り返すと初期値に依存せず、特定の値に収束することを2状態の遷移で具体的に確認する。
状態をa,bとし、初期の分布を$p_0 = [p_a, p_b]$、遷移確率を
$$
P = 
\begin{equation}
\begin{pmatrix}
\frac{1}{2} & \frac{1}{2} \\
\frac{1}{3} & \frac{2}{3} \\
\end{pmatrix}
\end{equation}
$$
とすると、1回の遷移後の状態、a,bの確率$p_1$は
$$
p_1 = p_0 P = 
\begin{equation}
\begin{pmatrix}
\frac{1}{2}p_a + \frac{1}{3}p_b & \frac{1}{2}p_a + \frac{2}{3}p_b \\
\end{pmatrix}
\end{equation}
$$
なお、上記のa,bの状態の確率を足すと、$\frac{1}{2}p_a + \frac{1}{3}p_b + \frac{1}{2}p_a + \frac{2}{3}p_b = p_a + p_b = 1$となっている。
Pに関して改めて考える。  
固有値は以下を考える。
$$
\begin{equation}
\begin{vmatrix}
\frac{1}{2} - \lambda & \frac{1}{2} \\
\frac{1}{3} & \frac{2}{3} - \lambda\\
\end{vmatrix}
\end{equation}
= {\lambda}^2 - \frac{7}{6}\lambda + \frac{1}{6} = 0
$$
よって、$\lambda = 1, \frac{1}{6}$。固有値は(右)固有ベクトル、左固有ベクトルでも同じ。
右固有ベクトルを考えると
$$
\begin{equation}
\begin{pmatrix}
1\\
1\\
\end{pmatrix}
(\lambda = 1),
\begin{pmatrix}
3\\
-2\\
\end{pmatrix}
\end{equation}
(\lambda = \frac{1}{6})
$$
左固有ベクトルを考えると、
$$
\begin{equation}
\begin{pmatrix}
2 & 3\\
\end{pmatrix}
(\lambda = 1),
\begin{pmatrix}
1 & -1\\
\end{pmatrix}
\end{equation}
(\lambda = \frac{1}{6})
$$
固有値、固有ベクトルをもとにPの対角化を考えると
$$
P = -\frac{1}{5}
\begin{equation}
\begin{pmatrix}
-2 & -3 \\
-1 & 1 \\
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & \frac{1}{6} \\
\end{pmatrix}
\begin{pmatrix}
1 & 3 \\
1 & -2 \\
\end{pmatrix}
\end{equation}
$$
$p_0$にPをk回作用させ、$k \rightarrow \infty$とすると、
$$
\lim_{k \to \infty} p_0 P^k = \lim_{k \to \infty} -\frac{1}{5}p_0
\begin{equation}
\begin{pmatrix}
-2 & -3 \\
-1 & 1 \\
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & {(\frac{1}{6})}^k \\
\end{pmatrix}
\begin{pmatrix}
1 & 3 \\
1 & -2 \\
\end{pmatrix}
= -\frac{1}{5} p_0
\begin{pmatrix}
-2 & -3 \\
-1 & 1 \\
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
0 & 0 \\
\end{pmatrix}
\begin{pmatrix}
1 & 3 \\
1 & -2 \\
\end{pmatrix}
= -\frac{1}{5}
\begin{pmatrix}
p_a & p_b \\
\end{pmatrix}
\begin{pmatrix}
-2 & -3 \\
-2 & -3 \\
\end{pmatrix}
=\frac{p_a+p_b}{5}
\begin{pmatrix}
2 & 3 \\
\end{pmatrix}
= 
\begin{pmatrix}
\frac{2}{5} & \frac{3}{5} \\
\end{pmatrix}
\end{equation}$$
よって、初期値$p_0$に依存しない。
