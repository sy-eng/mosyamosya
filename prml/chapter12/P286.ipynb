P286の(12.26)から、(12.30)を考える。この部分は演習12.3につながっている。

(12.1)より
$$
{\bf S} = \frac{1}{N} \sum_{n} ((x_n - \bar{x})(x_n - \bar{x})^{T}) \in R^{D \times D}
$$
また、教科書にあるように、
$$
{\bf X} = 
\begin{equation}
\begin{pmatrix}
(x_1 - \bar{x})^{T}\\
(x_2 - \bar{x})^{T}\\
\vdots\\
(x_n - \bar{x})^{T}
\end{pmatrix}
\end{equation} \in R^{N \times D}
$$
なので、
$$
{\bf X}^T = 
\begin{equation}
\begin{pmatrix}
(x_1 - \bar{x}) & (x_2 - \bar{x}) & \cdots & (x_n - \bar{x})\\
\end{pmatrix}
\end{equation} \in R^{D \times N}
$$
よって、内積の計算を参考に、
$$
{\bf S} = \frac{1}{N} {\bf X}^T {\bf X} = \sum_{n} ((x_n - \bar{x})(x_n - \bar{x})^{T})
$$

さて、(12.17)を考慮して、(12.26)を得る。
$$
{\bf S}u_i = \frac{1}{N} {\bf X}^T {\bf X} u_i = \lambda_i u_i
$$
これの真ん中の項と、右側の項に${\bf X}$をかけて、(12.27)を得る。
$$
\frac{1}{N} {\bf X} ({\bf X}^T {\bf X} u_i) = \frac{1}{N} ({\bf X} {\bf X}^T) {\bf X} u_i = \lambda_i {\bf X} u_i
$$
この式の真ん中の項と右側の項で$v_i = {\bf X} u_i$とすると、(12.28)を得る。
$$
\frac{1}{N} ({\bf X} {\bf X}^T) v_i = \lambda_i v_i
$$
この式は、${\bf A} = \frac{1}{N} ({\bf X} {\bf X}^T) \in R^{N \times N}$とすると、
Aの固有値を求める方程式となっている。かつ、${\bf S}$に対する、固有値は${\bf A}$に対する固有値と一致する。
更に、これに${\bf X}^T$をかけて、(12.29)を得る。
$$
\frac{1}{N} {\bf X}^T({\bf X} {\bf X}^T) v_i = \frac{1}{N} ({\bf X}^T {\bf X}) ({\bf X}^T v_i) = \lambda_i ({\bf X}^T v_i)
$$
この式より、$u_i \varpropto {\bf X}^T v_i$が言える。

(12.28)を利用し、
$$
v_i = \frac{1}{N\lambda_i}{\bf X}{\bf X}^T v_i
$$
$a u_i = {\bf X}^T v_i$として、
$\| u_i \| = 1, \| v_i \| = 1$とすると、
$$
1 = {\| v_i \|}^2 = {v_i}^T v_i = {v_i}^T (\frac{1}{N\lambda_i}{\bf X}{\bf X}^T v_i) = \frac{1}{N\lambda_i} {v_i}^T {\bf X}{\bf X}^T v_i
= \frac{1}{N\lambda_i} (a {u_i}^T)(a {u_i}) = \frac{a^2}{N\lambda_i} \| u_i \| = \frac{a^2}{N\lambda_i}
$$
両端の項の比較により、
$$
a = \frac{1}{{(N\lambda_i)}^\frac{1}{2}}
$$
(正負は入れ替わっても構わない)

よって、
$$
u_i = \frac{1}{{(N\lambda_i)}^\frac{1}{2}}{\bf X}^T v_i
$$

${\bf A}$について、考えてみる。
$$
n{\bf A} = 
\begin{equation}
\begin{pmatrix}
(x_1 - \bar{x})^{T}(x_1 - \bar{x}) & (x_1 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_1 - \bar{x})^{T}(x_n - \bar{x})\\
(x_2 - \bar{x})^{T}(x_1 - \bar{x}) & (x_2 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_2 - \bar{x})^{T}(x_n - \bar{x})\\
\vdots & \vdots & & \vdots\\
(x_n - \bar{x})^{T}(x_1 - \bar{x}) & (x_n - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_n - \bar{x})^{T}(x_n - \bar{x})\\
\end{pmatrix}
\end{equation}
$$
行列式を考える。最後の行にそれ以外の行を足すと、(12.1)より、$\sum_{m=1}^n(x_m - \bar{x}) = \sum_{m=1}^n x_m - n\bar{x} = 0$なので、
$$
n|{\bf A}| = 
\begin{equation}
\begin{vmatrix}
(x_1 - \bar{x})^{T}(x_1 - \bar{x}) & (x_1 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_1 - \bar{x})^{T}(x_n - \bar{x})\\
(x_2 - \bar{x})^{T}(x_1 - \bar{x}) & (x_2 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_2 - \bar{x})^{T}(x_n - \bar{x})\\
\vdots & \vdots & & \vdots\\
(x_n - \bar{x})^{T}(x_1 - \bar{x}) & (x_n - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_n - \bar{x})^{T}(x_n - \bar{x})\\
\end{vmatrix}
=
\begin{vmatrix}
(x_1 - \bar{x})^{T}(x_1 - \bar{x}) & (x_1 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_1 - \bar{x})^{T}(x_n - \bar{x})\\
(x_2 - \bar{x})^{T}(x_1 - \bar{x}) & (x_2 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_2 - \bar{x})^{T}(x_n - \bar{x})\\
\vdots & \vdots & & \vdots\\
(\sum_{m=1}^n(x_m - \bar{x}))^{T}(x_1 - \bar{x}) & (\sum_{m=1}^n(x_m - \bar{x}))^{T}(x_2 - \bar{x}) & \cdots & (\sum_{m=1}^n(x_m - \bar{x}))^{T}(x_n - \bar{x})\\
\end{vmatrix}
=
\begin{vmatrix}
(x_1 - \bar{x})^{T}(x_1 - \bar{x}) & (x_1 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_1 - \bar{x})^{T}(x_n - \bar{x})\\
(x_2 - \bar{x})^{T}(x_1 - \bar{x}) & (x_2 - \bar{x})^{T}(x_2 - \bar{x}) & \cdots & (x_2 - \bar{x})^{T}(x_n - \bar{x})\\
\vdots & \vdots & & \vdots\\
0 & 0 & \cdots & 0\\
\end{vmatrix}
= 0
\end{equation}
$$
つまり、${\bf A}$は正則でなく、Aの固有値は0を含む。

${\bf A}$の要素を考えると
$$
{\bf A}_{ij} = \frac{1}{N} (x_i - \bar{x})^{T}(x_j - \bar{x}) = \phi(x_i)\phi(x_j) = k(x_i, x_j)
$$
ここで、$\phi(x_i) = \frac{1}{\sqrt{N}}(x_i - \bar{x})$。  
すると、12.3のカーネル主成分分析のように見なせる。  
$M = D, {\bf K} = {\bf A}$と考えられる。
